### **1**. Моделирование дискретного прямоугольного сигнала $s_1(k)$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 32      # Длина последовательности
n = np.arange(0, N)  # Дискретное нормированное время
U = 18 # Амплитуда
n0 = 5 # Начальный момент импульса
n_imp = 7 # Длина импульса

# Функция дискретного единичного скачка в момент k
def sigma_d(k):
  return int(k >= 0)

# Функция дискретного прямоугольного импульса в момент k
def s1_k(k):
  return U*(sigma_d(k - n0) - sigma_d(k - (n0+n_imp)))


# Создание дискретного прямоугольного сигнала
s_1 = []
for k in range(N):
  s_1.append(s1_k(k))

# График дискретного прямоугольного импульса от нормированного времени n
plt.figure(figsize=(10,4))
plt.stem(n, s_1, linefmt='-', markerfmt='o', basefmt=' ')
plt.title('Дискретный прямоугльный сигнал $s_1(k)$')
plt.xlabel('Нормированное время n')
plt.ylabel('$s_1(k)$')
plt.grid(True)
```

В ходе выполнения данного пункта смоделирован дискретный прямоугольный импульс $s_1(k)$ длиной $N=32$ на интервале времени $n \in [0,N−1]$.

Данный сигнал был смоделирован с помощью дискретного единичного скачка $\sigma_d(k)$: использовалась разность двух скачков с задержкой $n_0$ и $n_0 + n_{imp}$ домноженная на амплитуду имульса $U$:

$s_1(k) = U* (\sigma_d(k-n_0) - \sigma_d(k-(n_0 + n_{imp})))$

Использование задержки позволяет ограничить импульс в исходных границах со сдвигом во времени.

### **2**. Моделирование линейной комбинации гармонических сигналов $s_2(k)$

```python
import numpy as np
import matplotlib.pyplot as plt
import math


N = 32      # Длина последовательности
n = np.arange(0, 5*N)  # Дискретное нормированное время
a = [3.5, 2.7, 3.4] # Коэффициенты линейной комбинации гармонических сигналов
B = [3.5,7.7,4.2] # Амплитуды гармонических сигналов
w = [math.pi/6, math.pi/10, math.pi/18] # Частоты гармонических сигналов

# Функция последовательности x_i
def x_i(i,k):
  return B[i]*np.sin(w[i]*k)

# Функция последовательности s_2
def s2_k(k):
  return a[0]*x_i(0,k)+a[1]*x_i(1,k)+a[2]*x_i(2,k)

# Моделирование линейной комбинации
s_2 = []
for k in range(0, 5*N):
  s_2.append(s2_k(k))

# Моделирование дискретных гармонических сигналов x_i
x_1 = [x_i(0,k) for k in range(5*N)]
x_2 = [x_i(1,k) for k in range(5*N)]
x_3 = [x_i(2,k) for k in range(5*N)]

# Построение графиков
fig, axes = plt.subplots(4, 1, figsize=(12, 10))
axes[0].stem(n, s_2, linefmt='-', markerfmt='o', basefmt=' ')
axes[0].set_title('Линейная комбинация гармонических сигналов $s_2(k)$')
axes[0].set_xlabel('Нормированное время n')
axes[0].set_ylabel('$s_2(k)$')
axes[0].grid(True)

axes[1].stem(n, x_1, linefmt='-', markerfmt='o', basefmt=' ')
axes[1].set_title('Гармонический сигнал $x_1(k)$')
axes[1].set_xlabel('Нормированное время n')
axes[1].set_ylabel('$x_1(k)$')
axes[1].grid(True)

axes[2].stem(n, x_2, linefmt='-', markerfmt='o', basefmt=' ')
axes[2].set_title('Гармонический сигнал $x_2(k)$')
axes[2].set_xlabel('Нормированное время n')
axes[2].set_ylabel('$x_1(k)$')
axes[2].grid(True)

axes[3].stem(n, x_3, linefmt='-', markerfmt='o', basefmt=' ')
axes[3].set_title('Гармонический сигнал $x_3(k)$')
axes[3].set_xlabel('Нормированное время n')
axes[3].set_ylabel('$x_1(k)$')
axes[3].grid(True)

plt.tight_layout()
plt.show()
```

```python
mean_s2 = sum(s_2)/ (5*N)
E = sum([s_i**2 for s_i in s_2])
P = E/(5*N)

print(f'Среднее значение: {mean_s2}')
print(f'Энергия: {E}')
print(f'Средняя можность: {P}')
```

В ходе выполнения данного пункта смоделирована линейная комбинация дискретных гармонических сигналов $s_2(k)$ длиной $5N=5*32 = 160$ на интервале времени $n \in [0,5*N−1]$.

Значение вычисленных характеристик для последовательности:

Среднее значение: $mean_{s2} \approx 1.16$

Энергия: $E \approx 62808.19$

Средняя можность: $P \approx 392.55$

a. При моделировании линейной комбинации дискретных гармонических сигналов $s_2(k)$ использовались три гармонических сигнала $x_1(k)$, $x_2(k)$, $x_3(k)$ с различными частотами $\hat{\omega}_i$ и амплитудами $B_i$, каждый из которых домножался на соответствующий весовой коэффициент $a_i$, после чего производилось суммирование, полученных сигналов.

b. Вычисленные характеристики результирующего сигнала $s_2(k)$ определяются следующим образом:

$mean_{s2} = \frac{1}{5N} \sum_{k=0}^{5N}s_2(k)$

$E = \sum_{k=0}^{5N}s_2^2(k)$

$P = \frac{E}{5N} = \frac{1}{5N} \sum_{k=0}^{5N}s_2^2(k)$



### **3**. Моделирование дискретного гармонического сигнала $s_3(k)$

```python
import numpy as np
import matplotlib.pyplot as plt
import math


N = 32      # Длина последовательности
n = np.arange(0, N)  # Дискретное нормированное время
a = -0.885 # Основание экспоненты
C = 3 # Амплитуда гармонического сигнала
w0 = math.pi/8 # Частота гармонического сигнала

# Функция дискретного гармонического сигнала в момент k
def s3_k(k):
  return C*(abs(a)**k)*math.cos(w0*k)

# Моделирование дискретного сигнала
s_3 = []
for k in range(N):
  s_3.append(s3_k(k))

# Моделирование экспоненциальной огибающей
a_k = [abs(a)**k for k in range(N)]
a_k_ = [-elem for elem in a_k]

# Построение графиков
plt.figure(figsize=(10,4))
plt.stem(n, s_3, linefmt='b-', markerfmt='bo', basefmt=' ', label='$s_3(k)$')
plt.plot(n, a_k, color='purple', linestyle='dashed', label='Верхняя огибающая')
plt.plot(n, a_k_, color='purple', linestyle='dashed', label='Нижняя огибающая')
plt.title('Дискретный гармонический сигнал $s_3(k)$')
plt.xlabel('Нормированное время n')
plt.ylabel('$s_3(k)$')
plt.legend(loc='upper right')
plt.grid(True)

```

В ходе выполнения данного пункта смоделирован дискретный гармонический сигнал $s_3(k)$ длиной $N=32$ на интервале времени $n \in [0,N−1]$.

Аналитическая формула сигнала:

$s_3(k) = |a|^k * C * cos(\hat{\omega}_0k)$

Дискретный гармонический сигнал $s_3(k)$ представляет собой произведение двух компонент: экспоненциальной функции $|a|^k$, определяющей закон изменения амплитуды во времени, и гармонического сигнала $s(k) = C * cos(\hat{\omega}_0k)$, задающего периодическую структуру колебаний. При |a| < 1 наблюдается эффект затухания амплитуды колебаний по экспоненциальному закону.

### **4**. Моделирование периодической последовательности $s_4(k)$ дискретных прямоугольных импульсов


```python
import numpy as np
import matplotlib.pyplot as plt

N = 32 # Длина последовательности
U = 18 # Амплитуда
n0 = 5 # Начальный момент импульса
n_imp = 7 # Длина импульса

T = 2*n_imp # Период последовательности
n = np.arange(0, 5*T)  # Дискретное нормированное время

# Функция дискретного единичного скачка в момент k
def sigma_d(k):
  return int(k >= 0)

# Функция дискретного прямоугольного импульса в момент k
def s1_k(k):
  return U*(sigma_d(k - n0) - sigma_d(k - (n0+n_imp)))

# Функция периодической последовательности в момент k
def s4_k(k):
  return s1_k(k % T)

# Моделирование периодической последовательности
s_4 = []
for k in range(5*T):
  s_4.append(s4_k(k))

# График периодической последовательности от нормированного времени n
plt.figure(figsize=(10,4))
plt.stem(n, s_4, linefmt='-', markerfmt='o', basefmt=' ')
plt.title('Периодическая последовательность $s_4(k)$')
plt.xlabel('Нормированное время n')
plt.ylabel('$s_4(k)$')
plt.grid(True)
```

В ходе выполнения данного пункта смоделирована периодическая последовательность дискретных прямоугольных импульсов $s_4(k)$. Отображено 5 периодов последовательности $T=2n_{imp}$.

Моделирование периодической последовательности $s_4(k)$ прямоугольных импульсов осуществляется путем формирования прямоугольного импульса длительностью $n_{imp}$ отсчетов из первого пункта лабораторной, обеспечивая переодичность этого импульса путем применения операции взятия остатка от деления $k \mod T$.
