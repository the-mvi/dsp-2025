##Основные теоретические положения
**Единичный цифровой импульс** (или дискретная дельта-функция, $\delta_d(n))$ — это базовая дискретная функция, которая принимает значение 1 в нулевом отсчёте и 0 во всех остальных.
$\delta_d(n) =
\begin{cases}
1, & n = 0,\\
0, & n \neq 0.
\end{cases}
\qquad n \in \mathbb{Z}$


Является дискретным аналогом дельта-функции
(функции Дирака):

$\delta(t) =
\begin{cases}
+\infty, & t = 0, \\
0, & t \neq 0
\end{cases}$ , $\int_{-\infty}^{+\infty} \delta(t) \, dt = 1$

**Дискретный единичный скачок** $\sigma_d(n)$ - базовая дискретная функция, которая принимает значение 0 в отсчетах, меньших нуля, и значение 1 во всех остальных.

$\sigma_d(n) =
\begin{cases}
1, & n \geq 0,\\
0, & n < 0.
\end{cases}
\qquad n \in \mathbb{Z}$

Является дискретным аналогом функции единичного скачка (функция Хэвисайда):

$\sigma(t) =
\begin{cases}
1, & t > 0 \\
0.5 & или & неопределена, & t = 0 \\
0, & t < 0 \\
\end{cases}$

**Дискретная экспоненциальная функция** задаётся формулой:
$
s(k) =
\begin{cases}
0, & k < 0, \\
a^k, & k \geq 0,
\end{cases}
\qquad k \in \mathbb{Z}.
$

**Свойства**
- Если $0<|a|<1$, последовательность является затухающей экспонентой.
- Если $|a|>1$, последовательность возрастает.
- Если $a=1$, то $s(k)=1$, то есть постоянная последовательность.

**Дискретный комплексный гармонический сигнал** задаётся формулой:
$ s_2(k) = C\exp (j \hat{\omega}_0 k)$,
где:
- $C$ — амплитуда гармонического сигнала,
- $\hat{\omega}_0$ — угловая частота дискретного гармонического сигнала,
- $k$ — индекс дискретного времени.

### **1**. Моделирование единичного цифрового импульса $\delta_d(k)$


```python
import numpy as np
import matplotlib.pyplot as plt

# длина последовательности
N = 32
# период дискретизации
T = 0.0015

# дискретные индексы
n = np.arange(N)           # 0 ... N-1
t = n * T                  # дискретное время

# единичный импульс
delta = np.zeros(N)
delta[0] = 1               # импульс в нуле

# график во времени nT
plt.figure(figsize=(10,4))
plt.stem(t, delta, basefmt=' ')
plt.title("Единичный цифровой импульс $\\delta_d(n)$ во времени nT")
plt.xlabel("Время t = nT (сек)")
plt.ylabel("Амплитуда")
plt.grid(True)
plt.show()

# график в нормированном времени n
plt.figure(figsize=(10,4))
plt.stem(n, delta, basefmt=' ')
plt.title("Единичный цифровой импульс $\\delta_d(n)$ в нормированном времени n")
plt.xlabel("Отсчёты n")
plt.ylabel("Амплитуда")
plt.grid(True)
plt.show()
```

В ходе выполнения задания был смоделирован дискретный сигнал $\delta_d(n)$ длиной $N=32$ с шагом дискретизации $T=0.0015 с$.

В нормированном дискретном времени $n$ импульс представлен «шпилькой» в нулевом отсчёте $(n = 0)$, остальные отсчёты равны нулю.

В дискретном времени $t=nT$ наблюдается та же форма импульса, но на физической временной оси. Это подчёркивает взаимосвязь между индексом отсчёта $n$ и реальным временем $t$: каждый отсчёт $n$ соответствует моменту времени $t=nT$.

1. Взаимосвязь дискретного и дискретного нормированного времени:

Нормированное время $n$ используется для анализа сигналов как последовательности чисел (индексов).

Дискретное время $t=nT$ позволяет интерпретировать сигналы в реальном физическом масштабе.

На практике оба представления эквивалентны: один задаёт «положение импульса» в индексах, другой — в реальном времени.

2. Различие между единичным цифровым импульсом и дельта-функцией Дирака:

Дельта-функция Дирака $\delta(t)$ — непрерывная функция, её амплитуда бесконечна в точке $t = 0$ и ноль в остальных точках, при этом интеграл равен 1.

Единичный цифровой импульс $\delta_d(n)$ — дискретная функция, принимающая конечные значения (0 или 1). Она является дискретным аналогом дельта-функции, но не обладает бесконечным пиком и определяется только в дискретных точках.


### **2**. Моделирование дискретного единичного скачка $\sigma_d(k)$


```python
import numpy as np
import matplotlib.pyplot as plt

N = 32      # Длина последовательности
T = 0.0015  # Период дискретизации

n = np.arange(0, N)  # Дискретное нормированное время
nt = n * T    # Дискретное время

# Создание дискретного единичного скачка
sigma_d = np.ones(N)

# Построение графиков
# График скачка от нормированного времени n
plt.figure(figsize=(10,4))
plt.stem(n, sigma_d, linefmt='-', markerfmt='o', basefmt=' ')
plt.title('Дискретный единичный скачок $\\sigma_d(k)$ в нормированном времени n')
plt.xlabel('Нормированное время n')
plt.ylabel('$\\sigma_d(k)$')
plt.grid(True)

# График скачка от дискретного времени nT
plt.figure(figsize=(10,4))
plt.stem(nt, sigma_d, linefmt='-', markerfmt='o', basefmt=' ')
plt.title('Дискретный единичный скачок $\\sigma_d(k)$ во времени nT')
plt.xlabel('Дискретное время nT')
plt.ylabel('$\\sigma_d(k)$')
plt.grid(True)
```

В ходе выполнения задания был смоделирован дискретный сигнал $\sigma_d(n)$ длиной $N=32$ с шагом дискретизации $T=0.0015 с$.

В нормированном времени $n$ и в дискретном времени $t=nT$ скачок представлен как последовательность значений, равных $1$, начиная с момента включения (при $n=0$ или $t=0$). До этого момента (если бы рассматривались
$n<0$) значения сигнала были бы равны $0$.

1. Соответствие между дискретным и аналоговым единичными скачками:

Дискретный единичный скачок $\sigma_d(n)$ является аналогом функции единичного скачка (функция Хэвисайда) $\sigma(t)$, полученная путем соотношения отсчета дискретного сигнала значению аналогового сигнала в момент времени $t=nT$:

$\sigma_d(n) = \sigma(nT)$

Аналоговый единичный скачок либо не определён в нуле, либо равен $0.5$, чем и отличается от дискретного.

2. Чему равна частота дискретизации дискретного единичного скачка?

Частота дискретизации дискретного единичного скачка равна $F_d = \frac{1}{T}$, где $T$ - период дискретизации.

Для $T = 0.0015$:

$F_d = \frac{1}{0.0015} \approx 666.6667 $

### 3. Моделирование дискретной экспоненциальной функции $ s_1(k) $

```python
import numpy as np
import matplotlib.pyplot as plt

a = -0.885
N = 32
T = 0.0015

# Дискретное нормированное время
n = np.arange(0, N)

# Дискретное реальное время
t = n * T

# Дискретная экспонента
s1 = a**n

# График
plt.figure(figsize=(10,4))
plt.stem(t, s1, basefmt=' ')
plt.title('Дискретная экспонента $s_1(nT)$')
plt.xlabel("Время t = nT (сек)")
plt.ylabel("Амплитуда")
plt.grid(True)

plt.figure(figsize=(10,4))
plt.stem(n, s1, basefmt=' ')
plt.title('Дискретная экспонента $s_1[n]$')
plt.xlabel("Отсчёты n")
plt.ylabel("Амплитуда")
plt.grid(True)

plt.show()


```

Моделирование показало корректное отображение дискретной экспоненты $s_1(n)$ как в нормированном времени, так и в дискретном $t=nT$.

Для $a=-0.885$ на графике получится чередующаяся кривая, которая быстро стремится к нулю по мере увеличения $n$.

Аналоговая экспонента:
$s(t) = e^{\alpha t}, \quad t \ge 0$

Дискретизация с шагом $T$ даёт:
$s[n] = s(nT) = e^{\alpha nT} = (e^{\alpha T})^n$

Таким образом, коэффициент $a$ в дискретной экспоненте соответствует
$a = e^{\alpha T}.$

Таким образом, дискретная экспонента $a^n$ является аналогом непрерывной экспоненты $e^{\alpha t}$,
а моделирование демонстрирует их взаимосвязь и корректность теоретических представлений.

### 4. Моделирование дискретного комплексного гармонического сигнала $\\s_2(k)$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 32      # Длина последовательности
T = 0.0015  # Период дискретизации

C = 3       # Амплитуда гармонического сигнала
w0 = np.pi / 8  #Частота гармонического сигнала

n = np.arange(N)
# Создание комплексного гармонического сигнала
s2_k = C * np.exp(1j * w0 * n)

# Разделение на вещественную и мнимую части
real_part = np.real(s2_k)
imag_part = np.imag(s2_k)

# Построение графиков
# График - вещественная часть
plt.figure(figsize=(10,4))
plt.stem(n, real_part, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.title('Вещественная часть дискретного комплексного гармонического сигнала: Re($s_2(k)$)')
plt.xlabel('Дискретное время n')
plt.ylabel('Амплитуда')
plt.grid(True)

# График - мнимая часть
plt.figure(figsize=(10,4))
plt.stem(n, imag_part, linefmt='g-', markerfmt='go', basefmt=' ')
plt.title('Мнимая часть комплексного дискретного гармонического сигнала: Im($s_2(k)$)')
plt.xlabel('Дискретное время n')
plt.ylabel('Амплитуда')
plt.grid(True)
```

В задании смоделирован дискретный комплексный гармонический сигнал $s_2(n)$ и выведены графики вещественной и мнимой части на интервале времени $n$.

Для записи дискретного комплексного гармонического сигнала как комбинацию двух вещественных последовательностей, можно использовать формулу Эйлера:

Формула Эйлера: $exp(j\alpha) = cos(\alpha) + j*sin(\alpha)$

Применяя к данному сигналу:

$s_2(k) = C*exp(j\omega_0k) = C*[cos(\omega_0k) + j*sin(\omega_0k)]$

Так как $C$ - вещественное:

$s_2(k) = C*cos(\omega_0k) + j*C*sin(\omega_0k) = 3*cos(\frac{\pi}{8}k) + j*3*sin(\frac{\pi}{8}k)$

Таким образом, получаем

Вещественная часть: $Re(s_2(k)) = C*cos(\omega_0k) = 3*cos(\frac{\pi}{8}k)$

Мнимая часть: $Im(s_2(k)) = C*sin(\omega_0k) = 3*sin(\frac{\pi}{8}k)$

Также исходя из графиков, можно отметить, что вещественная и мнимая часть имеет график функции $cos$ и $sin$ соответственно.

### 5. Графики последовательностей $ \delta_d(k)$, $\sigma_d(k)$, $s_1(k)$, задержанных на $m$ отсчетов.

1. Единичный цифровой импульс

Исходный импульс определяется как
$\delta_d(n) =
\begin{cases}
1, & n = 0, \\
0, & n \ne 0.
\end{cases}$

Задержанный на $m$ отсчётов импульс имеет вид:
$
\delta_{d,m}(n) = \delta_d(n-m) =
\begin{cases}
1, & n = m, \\
0, & n \ne m.
\end{cases}
$

При конечном интервале $n = 0,1,\dots,N-1$ это означает,
что импульс появляется в отсчёте $n=m$ (если $0 \le m \le N-1$),
иначе в рассматриваемом интервале он отсутствует.

2. Дискретный единичный скачок

Исходный скачок определяется как
$\sigma_d(n) =
\begin{cases}
1, & n \geq 0,\\
0, & n < 0.
\end{cases}$

Задержанный на $m$ отсчётов скачок имеет вид:
$
\sigma_{d,m}(n) = \sigma_d(n-m) =
\begin{cases}
1, & n \geq m, \\
0, & n < m.
\end{cases}
$

3. Дискретная экспоненциальная функция:

Исходная последовательность:
$
s_1(n) = a^n,
$

Задержанная версия:
$
s_{1,m}(n) = s_1(n-m) =
\begin{cases}
a^{\,n-m}, & n \ge m, \\[6pt]
0, & n < m.
\end{cases}
$

```python
import numpy as np
import matplotlib.pyplot as plt

N = 32
T = 0.0015
a = -0.885
C = 3
omega0 = np.pi / 8
m = 7

# Индексы и время
n = np.arange(N)          # 0..N-1
t = n * T

# Исходные последовательности
delta = np.zeros(N, dtype=float)
delta[0] = 1.0

s1 = a**n
# Задержанные последовательности
delta_m = np.zeros_like(delta)
if 0 <= m < N:
    delta_m[m] = 1.0

sigma_m = np.ones(N, dtype=float)
for i in range(m):
    sigma_m[i] = 0.0

s1_m = np.zeros_like(s1)
# для n >= m используем сдвиг (иначе остаётся 0)
idx = np.arange(m, N)
s1_m[idx] = a ** (idx - m)

# delta и delta_m
plt.figure(figsize=(10,4))
#plt.stem(n, delta, basefmt=" ", markerfmt="C0o", label=r'$\delta_d(n)$')
plt.stem(n, delta_m, basefmt=" ", markerfmt="C1s", label=fr'$\delta_d(n-{m})$')
plt.title(r'Единичный цифровой импульс $\delta_d(n)$ и задержанная версия')
plt.xlabel('n (отсчёты)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid(True)
plt.tight_layout()

# sigma и sigma_m
plt.figure(figsize=(10,4))
plt.stem(n, sigma_m, basefmt=" ", markerfmt="C1s", label=fr'$\sigma_d(n-{m})$')
plt.title(r'Задержанный дискретный единичный скачок $\sigma_{d,m}(n)$')
plt.xlabel('n (отсчёты)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid(True)
plt.tight_layout()

# s1_m
plt.figure(figsize=(10,4))
plt.stem(n, s1_m, basefmt=" ", markerfmt="C1s", label=fr'$s_1(n-{m})$')
plt.title(r'Дискретная экспоненциальная функция $s_1(n)$ и задержанная версия')
plt.xlabel('n (отсчёты)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

Моделирование задержки дискретных последовательностей показало, что задержка на $m$ отсчётов соответствует сдвигу сигнала вправо по оси дискретного времени.

Для импульса $\delta_d(n) $ это проявляется как смещение единичного отсчёта с позиции $n=0$ на позицию $n=m$.

Для скачка $\sigma_d(n) $ задержка приводит к смещению появления единиц с позиции $n=0$ на позицию $n=m$, а первые $m$ отсчётов заполняются нулями.

Для экспоненциальной функции $s_1(n)$ задержка приводит к появлению нулевых значений в первых $m$ отсчётах и повторению исходной формы сигнала, но сдвинутой во времени.

Основные результаты:  
- Показано, что нормированное время $n$ соответствует номеру отсчёта, а дискретное время $t=nT$ связывает сигнал с реальной временной шкалой. Таким образом, одна и та же последовательность может интерпретироваться в двух временных представлениях.  

- Единичный цифровой импульс $\delta_d(n)$ продемонстрировал отличие от дельта-функции: $\delta_d(n)$ существует только в дискретной области и принимает ненулевое значение лишь в одном отсчёте, в то время как непрерывная $\delta(t)$ является обобщённой функцией.  

- Дискретный единичный скачок $\sigma_d(n)$ является аналогом непрерывного единичного скачка, а его частота дискретизации - обратная величина к периоду $T$.  

- Дискретная экспоненциальная последовательность $s_1(n)$ подтвердила аналогию с непрерывной экспонентой $e^{\alpha t}$: при изменении основания экспоненты $a$ поведение сигнала меняется от затухающего к возрастающему, а знак $a$ влияет на чередование отсчётов.  

- Дискретный комплексный гармонический сигнал $s_2(n)$ был представлен в виде комбинации вещественной (косинус) и мнимой (синус) частей, что подтверждает фундаментальную связь между комплексной экспонентой и гармоническими колебаниями.  

- Исследование задержанных последовательностей показало, что операция задержки на $m$ отсчётов соответствует сдвигу сигнала вправо на $mT$ секунд во временной шкале. Для $\delta_d(n)$ это проявляется в смещении единичного импульса, для $\sigma_d(n)$ — в сдвиге момента начала скачка, для $s_1(n)$ — в появлении нулевых значений до момента задержки.

Таким образом, моделирование подтвердило корректность теоретических представлений о стандартных
дискретных сигналах, их взаимосвязи с непрерывными аналогами и особенностях обработки,
включая задержку и переход к комплексному представлению.
