### **1**. Вывод формул передаточных функций рекурсивных фильтров

**1.** Найдем формулу передаточной функции для формулы прямоугольников:

$$
y_{n+1} = y_n + s_{n+\frac{1}{2}}, y_0 = 0
$$

Пусть $s_n = e^{i\omega n}$ и $y_n = H(\omega)e^{i\omega n}$

Тогда получим:

$$
y_{n+1} = H(\omega)e^{i\omega (n+1)}
$$

$$
y_{n+1} = H(\omega)e^{i\omega (n)} +  e^{i\omega (n+\frac{1}{2})}
$$

Откуда, приравняв:

$$
H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega (n)} +  e^{i\omega (n+\frac{1}{2})}
$$

$$
H(\omega)e^{i\omega (n+1)} - H(\omega)e^{i\omega (n)} = e^{i\omega (n+\frac{1}{2})}
$$

$$
H(\omega)(e^{i\omega (n+1)} - e^{i\omega (n)}) = e^{i\omega (n+\frac{1}{2})}
$$

$$
H(\omega)e^{i\omega (n)}(e^{i\omega} - 1) = e^{i\omega (n)}e^{i\omega (\frac{1}{2})}
$$

$$
H(\omega)(e^{i\omega} - 1) = e^{i\omega (\frac{1}{2})}
$$

$$
H(\omega) = \frac{e^{i\omega \frac{1}{2}}}{e^{i\omega} - 1}=\frac{e^{i\omega \frac{1}{2}}}{e^{i\omega\frac{1}{2}}e^{i\omega\frac{1}{2}} - e^{i\omega\frac{1}{2}}e^{i\omega-\frac{1}{2}}} = \frac{1}{e^{i\omega\frac{1}{2}} - e^{i\omega-\frac{1}{2}}} = \frac{1}{cos(\frac{\omega}{2})+isin(\frac{\omega}{2}) - (cos(\frac{\omega}{2})-isin(\frac{\omega}{2}))} =\frac{1}{2isin(\frac{\omega}{2})}
$$

Точное значение интеграла от $e^{i\omega t}$ равно $\frac{e^{i\omega t}}{i\omega}$

Тогда отношение значений $\gamma$ будет равно:

$$
\gamma = \frac{Вычисленное}{Точное} = \frac{e^{i\omega \frac{1}{2}}e^{i\omega n}}{e^{i\omega} - 1} * \frac{i\omega}{e^{i\omega n}} = \frac{i\omega}{2isin(\frac{\omega}{2})} = \frac{\frac{\omega}{2}}{sin(\frac{\omega}{2})}
$$

**2.** Найдем формулу передаточной функции для формулы трапеций:

$$
y_{n+1} = y_n + \frac{1}{2}(s_n+s_{n+1}), y_0 = 0
$$

Пусть $s_n = e^{i\omega n}$ и $y_n = H(\omega)e^{i\omega n}$

Тогда получим:

$$
y_{n+1} = H(\omega)e^{i\omega (n+1)}
$$

$$
y_{n+1} = H(\omega)e^{i\omega (n)} +  \frac{1}{2}(e^{i\omega(n)} + e^{i\omega(n+1)})
$$

Откуда, приравняв:

$$
H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega (n)} +  \frac{1}{2}(e^{i\omega(n)} + e^{i\omega(n+1)})
$$

$$
H(\omega)e^{i\omega (n+1)} - H(\omega)e^{i\omega (n)} = \frac{1}{2}(e^{i\omega(n)} + e^{i\omega(n+1)})
$$

$$
H(\omega)e^{i\omega (n)}(e^{i\omega} - 1) = \frac{1}{2}e^{i\omega(n)} ( 1 + e^{i\omega})
$$

$$
H(\omega)(e^{i\omega} - 1) = \frac{1}{2}(1 + e^{i\omega})
$$

$$
H(\omega) = \frac{1 + e^{i\omega}}{2(e^{i\omega} - 1)} = \frac{e^{i\omega\frac{1}{2}}e^{i\omega\frac{1}{2}} + e^{i\omega\frac{1}{2}}e^{i\omega-\frac{1}{2}}}{2(e^{i\omega\frac{1}{2}}e^{i\omega\frac{1}{2}} - e^{i\omega\frac{1}{2}}e^{i\omega-\frac{1}{2}})} =\frac{e^{i\omega\frac{1}{2}}+ e^{i\omega-\frac{1}{2}}}{e^{i\omega\frac{1}{2}}- e^{i\omega-\frac{1}{2}}} =\frac{cos(\frac{\omega}{2})}{2isin(\frac{\omega}{2})}
$$

Точное значение интеграла от $e^{i\omega t}$ равно $\frac{e^{i\omega t}}{i\omega}$

Тогда отношение значений $\gamma$ будет равно:

$$
\gamma = \frac{Вычисленное}{Точное} = \frac{cos(\frac{\omega}{2})*e^{i\omega n}}{2isin(\frac{\omega}{2})} * \frac{i\omega}{e^{i\omega n}} = \frac{cos(\frac{\omega}{2})*(i\omega)}{2isin(\frac{\omega}{2})} = \frac{cos(\frac{\omega}{2})*\frac{\omega}{2}}{sin(\frac{\omega}{2})}
$$

**3.** Найдем формулу передаточной функции для формулы Симпсона:

$$
y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1}+4s_{n}+s_{n+1}), y_0 = 0
$$

Пусть $s_n = e^{i\omega n}$ и $y_n = H(\omega)e^{i\omega n}$

Тогда получим:

$$
y_{n+1} = H(\omega)e^{i\omega (n+1)}
$$

$$
y_{n+1} = H(\omega)e^{i\omega (n-1)} +  \frac{1}{3}(e^{i\omega(n-1)}+4e^{i\omega(n)} + e^{i\omega(n+1)})
$$

Откуда, приравняв:

$$
H(\omega)e^{i\omega (n+1)} =H(\omega)e^{i\omega (n-1)} +  \frac{1}{3}(e^{i\omega(n-1)}+4e^{i\omega(n)} + e^{i\omega(n+1)})
$$

$$
H(\omega)e^{i\omega (n+1)} - H(\omega)e^{i\omega (n-1)} = \frac{1}{3}(e^{i\omega(n-1)}+4e^{i\omega(n)} + e^{i\omega(n+1)})
$$

$$
H(\omega)e^{i\omega (n)}(e^{i\omega} - e^{-i\omega}) = \frac{1}{3}e^{i\omega(n)} ( e^{-i\omega} +4 + e^{i\omega})
$$

$$
H(\omega)(e^{i\omega} - e^{-i\omega}) = \frac{(e^{-i\omega} +4 + e^{i\omega})}{3}
$$

$$
H(\omega) = \frac{( e^{-i\omega} +4 + e^{i\omega})}{3(e^{i\omega} - e^{-i\omega})} = \frac{cos(\omega) - isin(\omega)+ 4 + cos(\omega) + isin(\omega)}{3(cos(\omega) + isin(\omega)- (cos(\omega) - isin(\omega)))} = \frac{2cos(\omega)+ 4}{3(2isin(\omega))} = \frac{cos(\omega)+ 2}{3isin(\omega)}
$$

Точное значение интеграла от $e^{i\omega t}$ равно $\frac{e^{i\omega t}}{i\omega}$

Тогда отношение значений $\gamma$ будет равно:

$$
\gamma = \frac{Вычисленное}{Точное} = \frac{(cos(\omega)+ 2)*e^{i\omega n}}{3isin(\omega)} * \frac{i\omega}{e^{i\omega n}} = \frac{(cos(\omega)+ 2)*(i\omega)}{3isin(\omega)} =  \frac{(cos(\omega)+ 2)*\omega}{3sin(\omega)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def H_rectangle(w):
  return 1/(2*1j*np.sin(w/2))

def H_trapeze(w):
  return np.cos(w/2)/(2*1j*np.sin(w/2))

def H_simpson(w):
  return (np.cos(w)+2)/(3*1j*np.sin(w))


N = 4098
# Циклические частоты
F = [f/N for f in range(1, N)][:N//2 - 1]
# Круговые частоты
W = [2*np.pi * f for f in F ]

#Вычисление передаточных функций
h_rect = [H_rectangle(w) for w in W]
h_trapeze = [H_trapeze(w) for w in W]
h_simpson = [H_simpson(w) for w in W]

#Построение графиков
plt.figure()
plt.plot(F, [h.imag for h in h_rect], label='$H_{прямоугольников}$')
plt.plot(F, [h.imag for h in h_trapeze], label='$H_{трапеций}$')
plt.plot(F, [h.imag for h in h_simpson], label='$H_{Симпсона}$')

plt.xlabel('f')
plt.ylabel('$\\tilde H(f)$')
plt.title("График передаточных функций")
plt.ylim((-5,1))
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def g_rectangle(w):
  return (w/2)/(np.sin(w/2))

def g_trapeze(w):
  return np.cos(w/2)*(w/2)/(np.sin(w/2))

def g_simpson(w):
  return (np.cos(w)+2)*w/(3*np.sin(w))


N = 4098
# Циклические частоты
F = [f/N for f in range(1, N)][:N//2 - 1]
# Круговые частоты
W = [2*np.pi * f for f in F ]

#Вычисление отношений передаточных функций
g_rect = [g_rectangle(w) for w in W]
g_trapeze = [g_trapeze(w) for w in W]
g_simpson = [g_simpson(w) for w in W]

#Построение графиков
plt.figure()
plt.plot(F, g_rect, label='$\\gamma_{прямоугольников}$')
plt.plot(F, g_trapeze, label='$\\gamma_{трапеций}$')
plt.plot(F, g_simpson, label='$\\gamma_{Симпсона}$')

plt.xlabel('f')
plt.ylabel('$\\gamma$')
plt.title("График отношения значений $\\gamma$ передаточных функций")
plt.ylim((-1,5))
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

Интерпретация частотных свойств и поведение на малых частотах:

1. Все $H(\omega)$ имеют вид «что-то $ \div i\sin(\cdot)$», поэтому сами по себе $H(\omega)$ содержат фактор $1/\sin(\cdot)\sim 1/\omega$ при $\omega\to0$ — это естественно для интегрирующего поведения (интеграл даёт множитель $1/(i\omega)$). Поэтому для корректного сравнения мы ввели $\gamma(\omega)$ — отношение вычисленного значения к точному интегралу; для идеального интегратора $\gamma(\omega)\to1$ при $\omega\to0$.


2. Малочастотные (разложение в ряд по $\omega$) поведение $\gamma(\omega)$:

   — Прямоугольники:
   $$
   \gamma_{\text{rect}}(\omega)=1+\frac{\omega^2}{24}+\frac{7\omega^4}{5760}+\cdots
   $$
   То есть ошибка начинается с члена $O(\omega^2)$ и положительна (переоценка интеграла при малых, но ненулевых $\omega$).

   — Трапеции:
   $$
   \gamma_{\text{trap}}(\omega)=1-\frac{\omega^2}{12}-\frac{\omega^4}{720}+\cdots
   $$
   Ошибка тоже $O(\omega^2)$, но с противоположным знаком (недооценка).

   — Симпсон:
   $$
   \gamma_{\text{Simp}}(\omega)=1+\frac{\omega^4}{180}+\cdots
   $$
   У Симпсона отсутствует член $O(\omega^2)$ — первая нетривиальная ошибка $O(\omega^4)$. Это формальное подтверждение того, что для медленно меняющихся гармоник (малые $\omega$) схема Симпсона значительно точнее: погрешность растёт намного медленнее с частотой.

3. По амплитудно-частотной характеристике (в терминах $\gamma$):
   — Для $\omega$ близких к нулю все схемы стремятся к 1, но порядок аппроксимации разный: Симпсон — «наиболее плоская» (лучше аппроксимирует интеграл на низких частотах), затем идут прямоугольники и трапеции (оба $O(\omega^2)$, но с разными знаками и коэффициентами).

4. По фазе: поскольку $H(\omega)$ имеют основной множитель $1/i$ и дополнительные чисто вещественные множители (например $\cos(\cdot)$), фазовые искажения небольшие на низких частотах, но на высоких частотах начнут доминировать нули/полюса в числителях/знаменателях (особенно там, где $\sin$ в знаменателе приближается к нулю).

Таким образом, если нужно интегрировать медленно меняющий сигнал (низкочастотную составляющую) — лучше применять формулу Симпсона (она даёт порядок точности выше). Трапеция/прямоугольник проще и вычислительно дешевле, но точность хуже; знак второй порядка члена объясняет, почему одна схема систематически недооценивает, другая — переоценивает интеграл для малых гармоник.


### **2**. Вывод формулы передаточной функции рекурсивного фильтра для интегрирования по «правилу 3/8»

Найдем формулу для передаточной функции рекурсивного фильтра для интегрирования по "правилу 3/8":

$$
y_{n+2}=y_{n−1}+\frac{1}{8}(x_{n+2}+3x_{n+1}+3x_n+x_{n−1})
$$

Пусть $x_n = e^{i\omega n}$ и $y_n = H(\omega)e^{i\omega n}$

Тогда получим:

$$
H(\omega)e^{i\omega (n+2)} = H(\omega)e^{i\omega (n-1)} + \frac{1}{8}(e^{i\omega (n+2)}+3e^{i\omega (n+1)}+3e^{i\omega n}+e^{i\omega (n-1)})
$$

$$
H(\omega)e^{i\omega (n+2)} - H(\omega)e^{i\omega (n-1)} =\frac{1}{8}(e^{i\omega (n+2)}+3e^{i\omega (n+1)}+3e^{i\omega n}+e^{i\omega (n-1)})
$$

$$
H(\omega)e^{i\omega n}(e^{i2\omega}  - e^{-i\omega}) =\frac{1}{8}e^{i\omega n}(e^{i2\omega}+3e^{i\omega}+3+e^{-i\omega})
$$

$$
H(\omega)(e^{i2\omega}  - e^{-i\omega}) =\frac{1}{8}(e^{i2\omega}+3e^{i\omega}+3+e^{-i\omega})
$$

$$
H(\omega) =\frac{e^{i2\omega}+3e^{i\omega}+3+e^{-i\omega}}{8(e^{i2\omega}  - e^{-i\omega})} = \frac{e^{i\frac{\omega}{2}}(e^{i\frac{3\omega}{2}}+3e^{i\frac{\omega}{2}}+3e^{-i\frac{\omega}{2}}+e^{-i\frac{3\omega}{2}})}{8e^{i\frac{\omega}{2}}(e^{i\frac{3\omega}{2}} - e^{-i\frac{3\omega}{2}})} = \frac{cos(\frac{3\omega}{2})+ i sin(\frac{3\omega}{2})+3cos(\frac{\omega}{2})+3isin(\frac{\omega}{2})+3cos(\frac{\omega}{2})-3isin(\frac{\omega}{2})+cos(\frac{3\omega}{2})- isin(\frac{3\omega}{2})}{8(cos(\frac{3\omega}{2})+ i sin(\frac{3\omega}{2})-(cos(\frac{3\omega}{2})- isin(\frac{3\omega}{2})))} = \frac{2cos(\frac{3\omega}{2})+6cos(\frac{\omega}{2})}{8*2isin(\frac{3\omega}{2}))} = \frac{cos(\frac{3\omega}{2})+3cos(\frac{\omega}{2})}{8isin(\frac{3\omega}{2}))}
$$

Точное значение интеграла от $e^{i\omega t}$ равно $\frac{e^{i\omega t}}{i\omega}$

Тогда отношение значений $\gamma$ будет равно:

$$
\gamma = \frac{Вычисленное}{Точное} =\frac{(cos(\frac{3\omega}{2})+3cos(\frac{\omega}{2}))*e^{i\omega n}}{8isin(\frac{3\omega}{2}))} * \frac{i\omega}{e^{i\omega n}} = \frac{(cos(\frac{3\omega}{2})+3cos(\frac{\omega}{2}))*\omega}{8sin(\frac{3\omega}{2}))}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

def H_(w):
  return (np.cos(3*w/2)+3*np.cos(w/2))/(8*1j*np.sin(3*w/2))


N = 4098
# Циклические частоты
F = [f/N for f in range(1, N)][:N//2 - 1]
# Круговые частоты
W = [2*np.pi * f for f in F ]

#Вычисление передаточной функции
h_3_8 = [H_(w) for w in W]

#Построение графика
plt.figure()
plt.plot(F, [h.imag for h in h_3_8], label='$H_{3/8}$')

plt.xlabel('f')
plt.ylabel('$\\tilde H(f)$')
plt.title("График передаточной функции для интегрирования по 'правилу $\\frac{3}{8}$'")
plt.ylim((-10, 10))
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def g_(w):
  return (np.cos(3*w/2)+3*np.cos(w/2))*w/(8*np.sin(3*w/2))


N = 4098
# Циклические частоты
F = [f/N for f in range(1, N)][:N//2 - 1]
# Круговые частоты
W = [2*np.pi * f for f in F ]

#Вычисление отношения передаточной функции
g_3_8 = [g_(w) for w in W]

#Построение графика
plt.figure()
plt.plot(F, g_3_8, label='$\\gamma_{3/8}$')

plt.xlabel('f')
plt.ylabel('$\\gamma$')
plt.title("График отношения $\\gamma$ передаточной функции для интегрирования по 'правилу $\\frac{3}{8}$'")
plt.ylim((-10, 10))
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

1. Поведение при $\omega\to0$: подставляя малые $\omega$,
   $$
   \gamma_{3/8}(0)=\frac{4\cdot 0}{8\cdot(3/2)0}=\frac{4}{12}=\frac{1}{3}.
   $$
   То есть реализованная рекурсивная схема в форме имеет постоянный коэффициент $(=1/3)$ вместо единицы. Это значит: как интегратор она отдает на постоянной составляющей в 3 раза меньше, чем «идеальный» непрерывный интеграл. Чтобы восстановить корректное соответствие со «точным интегралом», требуется домножить выход на коэффициент $3$ (нормировка).

2. После нормировки (умножения на 3) малочастотное разложение:
   $$
   \big(3\gamma_{3/8}(\omega)\big)=1+\frac{3\omega^{4}}{240}+\cdots = 1 + \frac{\omega^{4}}{80}+\cdots
   $$
   т.е. первый ненулевой член — $O(\omega^4)$. Следовательно, после нормировки правило $3/8$ имеет ту же порядокную точность по частоте, что и правило Симпсона (оба — точны до членов $O(\omega^3)$ в полиномиальном смысле), но коэффициенты при $\omega^4$ отличаются: для Симпсона коэффициент при $\omega^4$ = $1/180\approx0.00556$, а для $3/8$ (после домножения на 3) — $1/80=0.0125$. То есть при малых ненулевых частотах Симпсон аппроксимирует интеграл точнее (меньший коэффициент $\omega^4$) чем нормированный (3/8).

3. Частотные особенности: знаменатель содержит $\sin(3\omega/2)$ — это дает нули/особенности в ответе в частотах, где $3\omega/2$ кратно $\pi$ (то есть $\omega = 2\pi m/3$). Это определяет расположение частот, где ответ резко меняется (пики/нулевые места), и важно учитывать при анализе спектра входного сигнала.

Правило $3/8$ при корректной нормировке обеспечивает четвертый порядок малочастотной аппроксимации (как и Симпсон), но с большим коэффициентом при $\omega^4$ — т.е. на малых частотах оно хуже Симпсона. Кроме того, у $3/8$ структура частотных особенностей (позиции нулей/особых частот) сдвинута из-за множителя $3$ в аргументе синуса, что дает отличную «полосовую» картину на средних/высоких частотах.

**3.** **Cопоставительный анализ частотных характеристик передаточных функций, полученных при выполнении п. 1 и 2.**

```python
import numpy as np
import matplotlib.pyplot as plt

N = 4098
F = np.array([f/N for f in range(1, N)])[:N//2-1]
W = 2*np.pi*F

# Функции gamma
def g_rectangle(w):
    return (w/2)/(np.sin(w/2))

def g_trapeze(w):
    return np.cos(w/2)*(w/2)/(np.sin(w/2))

def g_simpson(w):
    return (np.cos(w)+2)*w/(3*np.sin(w))

def g_3_8(w):
    return 3*(np.cos(3*w/2)+3*np.cos(w/2))*w/(8*np.sin(3*w/2))

# Вычисление
g_rect = g_rectangle(W)
g_trap = g_trapeze(W)
g_simp = g_simpson(W)
g_38 = g_3_8(W)

# Построение графика
plt.figure(figsize=(10,6))
plt.plot(F, g_rect, label='Прямоугольники', color='blue')
plt.plot(F, g_trap, label='Трапеции', color='green')
plt.plot(F, g_simp, label='Симпсон', color='red')
plt.plot(F, g_38, label='Правило 3/8', color='purple')
plt.xlabel('f')
plt.ylabel(r'$\gamma(\omega)$')
plt.title('Сравнительный анализ частотных характеристик интегрирующих фильтров')
plt.ylim(0, 2)
plt.grid(True)
plt.legend()
plt.show()
```

1. Малые частоты $\omega \to 0$:

   * Симпсон и правило 3/8 дают $\gamma \to 1$ после нормировки — наилучшая точность.
   * Прямоугольник: $\gamma \approx 1 + \omega^2/24$ — небольшая переоценка.
   * Трапеция: $\gamma \approx 1 - \omega^2/12$ — небольшая недооценка.
   Таким образом, симпсон более плоская кривая — меньше ошибка для низкочастотных гармоник.

2. Средние и высокие частоты:

   * Прямоугольник и трапеция начинают быстро расти/падать (амплитуда отклоняется от 1) из-за  $ \sin(\omega/2) $ в знаменателе.
   * Симпсон более устойчив, но всe же имеет пики/провалы при $\omega = n\pi$ (нули синуса в знаменателе).
   * 3/8 показывает характерные пики/провалы при $\omega = 2\pi n / 3$ (из-за $\sin(3\omega/2)$) — частотная структура отличается от Симпсона.

3. Выводы по точности:

   * Лучшие для низких частот: Симпсон > 3/8 (после нормировки) > трапеции ≈ прямоугольники.
   * Для средних и высоких частот: правило 3/8 имеет более «сильные» резонансы/нули из-за множителя 3, Симпсон более плавная.
   * Применение: для медленно меняющихся сигналов лучше Симпсон; для более частотных спектров — учитывать пики у 3/8.

### **4.** **Выводы**

В работе исследованы частотные свойства рекурсивных фильтров для интегрирования дискретного сигнала по различным квадратурным формулам: прямоугольников, трапеций, Симпсона и правила 3/8.

* На низких частотах формулы Симпсона и 3/8 (после нормировки) дают высокую точность $O(\omega^4)$, прямоугольники и трапеции — более грубую $O(\omega^2)$.
* Симпсон обеспечивает наиболее плоскую амплитудно-частотную характеристику, правило 3/8 имеет характерные пики на средних/высоких частотах.
* Для медленно меняющихся сигналов оптимальна схема Симпсона; прямоугольники и трапеции подходят для менее точного интегрирования.
