# Основные теоретические положения

## 1. Дискретное время и нормирование

### Дискретное время
- **Дискретное время**: $nT$, где:
  - $n = 0, 1, 2, \ldots, N-1$ - целочисленный индекс отсчёта
  - $T$ - период дискретизации (интервал между отсчётами в секундах)
- **Нормированное время**: $n$ (безразмерная величина)
- **Связь**: $n_{\text{норм}} = \dfrac{nT}{T} = n$

## 2. Базовые дискретные сигналы

### Единичный цифровой импульс
$$
\delta_d(k) =
\begin{cases}
1, & k = 0 \\
0, & k \neq 0
\end{cases}
$$

**Обозначения:**
- $k$ - дискретное время (целочисленный аргумент)
- $\delta_d(k)$ - значение импульса в момент $k$

### Дискретный единичный скачок
$$
\sigma_d(k) =
\begin{cases}
1, & k \geq 0 \\
0, & k < 0
\end{cases}
$$

**Обозначения:**
- $\sigma_d(k)$ - значение скачка в момент $k$

## 3. Смещённые и составные сигналы

### Дискретный единичный скачок, смещенный на $t_0$ отсчетов
$$
\sigma_d(k - t_0) =
\begin{cases}
1, & k \geq t_0 \\
0, & k < t_0
\end{cases}
$$

**Обозначения:**
- $t_0$ - величина сдвига (целое число)
- $\sigma_d(k - t_0)$ - скачок, начинающийся в момент $t_0$

### Прямоугольный импульс через скачки
$$
s_1(k) = U \cdot [\sigma_d(k - n_0) - \sigma_d(k - (n_0 + n_{imp}))]
$$

**Обозначения:**
- $U$ - амплитуда импульса
- $n_0$ - начальный момент импульса
- $n_{imp}$ - длительность импульса
- $\sigma_d(k - n_0)$ - "включение" импульса
- $\sigma_d(k - (n_0 + n_{imp}))$ - "выключение" импульса

**Аналитически:**
$$
s_1(k) =
\begin{cases}
U, & n_0 \leq k \leq n_0 + n_{imp} - 1 \\
0, & \text{иначе}
\end{cases}
$$

## 4. Экспоненциальные и гармонические сигналы

### Дискретная экспоненциальная функция
$$
s(k) =
\begin{cases}
0, & k < 0 \\
a^k, & k \geq 0
\end{cases}
$$

**Обозначения:**
- $a$ - основание экспоненты (действительное или комплексное число)
- $k$ - дискретное время

**Типы поведения:**
- $|a| > 1$ - возрастающая экспонента
- $|a| < 1$ - затухающая экспонента  
- $|a| = 1$ - постоянная амплитуда
- $a < 0$ - знакопеременная последовательность

### Гармонический сигнал с экспоненциальной огибающей
$$
s_3(k) = C \cdot \cos(\omega_0 k) \cdot |a|^k
$$

**Обозначения:**
- $C$ - амплитуда гармонического сигнала
- $\omega_0$ - нормированная круговая частота (рад/отсчёт)
- $a$ - основание экспоненциальной огибающей
- $|a|^k$ - экспоненциальная огибающая

Моделирует затухающие или нарастающие колебания.

## 5. Линейные комбинации сигналов

### Принцип суперпозиции
$$
s(k) = \sum_{i=1}^{n} a_i x_i(k)
$$

**Обозначения:**
- $x_i(k)$ - элементарные сигналы-компоненты
- $a_i$ - весовые коэффициенты
- $s(k)$ - результирующий сигнал

### Гармонические компоненты
$$
x_i(k) = B_i \sin(\omega_i k)
$$

**Обозначения:**
- $B_i$ - амплитуда $i$-й гармоники
- $\omega_i$ - нормированная круговая частота $i$-й гармоники

## 6. Частотно-временные соотношения

### Связь частоты и периода гармонического сигнала
$$
\hat{\omega} = \frac{2\pi}{T} \Leftrightarrow T = \frac{2\pi}{\hat{\omega}}
$$

**Обозначения:**
- $\hat{\omega}$ - нормированная круговая частота (рад/отсчёт)
- $T$ - период в отсчётах

**Условие периодичности:** Сигнал $x(k) = \sin(\hat{\omega}k)$ периодичен если $\frac{2\pi}{\hat{\omega}}$ - рациональное число.

## 7. Энергетические характеристики

### Среднее значение сигнала
$$
\overline{s(k)} = \frac{1}{L} \sum_{k=0}^{L-1} s(k)
$$

### Энергия сигнала
$$
E = \sum_{k=0}^{L-1} |s(k)|^2
$$

### Средняя мощность
$$
P = \frac{1}{L} \sum_{k=0}^{L-1} |s(k)|^2 = \frac{E}{L}
$$

## 8. Периодические последовательности

### Условие периодичности
$$
s(k + T) = s(k) \quad \forall k
$$

**Обозначения:**
- $T$ - период последовательности (целое число)
- $s(k)$ - периодический сигнал

### Прямоугольная периодическая последовательность
$$
s_4(k) =
\begin{cases}
U, & k \mod T < n_{imp} \\
0, & \text{иначе}
\end{cases}
$$

**Обозначения:**
- $U$ - амплитуда импульсов
- $T$ - период повторения
- $n_{imp}$ - длительность импульса в отсчётах
- $k \mod T$ - остаток от деления $k$ на $T$

## Задание 1

Смоделировать дискретный прямоугольный импульс:

$ s_1(k) = \begin{cases} U, & \quad n_0 \leq k \leq n_0 + n_{imp} - 1; \\ 0, & \quad \text{иначе}; \end{cases} $ \
на основе дискретного единичного скачка $\sigma_d(k)$ с выводом графика на интервале времени $n \in [0, N - 1]$. Пояснить, как выполняется моделирование импульса.

```python
# Вектор дискретного времени k (k_1, длина N)
k_1 = np.arange(N)

# Моделирование единичного скачка sigma_d(k - n_сдвига)
# Скачок, начинающийся в n0: 1 для k >= n0, 0 иначе
sigma_n0 = np.where(k_1 >= n0, 1, 0)

# Скачок, начинающийся в n0 + n_imp: 1 для k >= n0 + n_imp, 0 иначе
end_index = n0 + n_imp
sigma_end = np.where(k_1 >= end_index, 1, 0)

# s1(k) = U * [sigma_d(k - n0) - sigma_d(k - (n0 + n_imp))]
s1_k = U * (sigma_n0 - sigma_end)

# Визуализация s1(k)
plt.figure(figsize=(10, 4))
plt.stem(k_1, s1_k, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.title(f'Дискретный прямоугольный импульс $s_1(k)$')
plt.xlabel('Дискретное время $k$')
plt.ylabel('$s_1(k)$')
plt.grid(True, linestyle=':', alpha=0.7)
plt.xticks(np.arange(0, N, 2))
plt.ylim(-1, U + 1)
plt.show()
```


**Моделирование выполняется с использованием дискретного единичного скачка $\sigma_d(k)$:**
$$
\sigma_d(k) =
\begin{cases}
1, & k \geq 0 \\
0, & k < 0
\end{cases}
$$

Прямоугольный импульс можно представить как разность двух единичных скачков:

$$
s_1(k) = U \cdot [\sigma_d(k - n_0) - \sigma_d(k - (n_0 + n_{imp}))]
$$

где:
- $U$ - амплитуда импульса
- $n_0$ - начальный момент импульса
- $n_{imp}$ - длительность импульса
- $\sigma_d(k)$ - дискретный единичный скачок

**Процесс моделирования** строится по следующему алгоритму: сначала создаётся единичный скачок, начинающийся в момент $n_0$, затем создаётся второй единичный скачок, начинающийся в момент $n_0 + n_{imp}$. После этого второй скачок вычитается из первого, и полученный результат умножается на амплитуду $U$.

## Задание 2

Смоделировать линейную комбинацию дискретных гармонических сигналов $ s_2(k) $:

$ s_2(k) = a_1 x_1(k) + a_2 x_2(k) + a_3 x_3(k) $,

где


$ x_i(k) = B_i \sin(\hat{\omega_i}k), \quad i = 1,2,3 $

с выводом графиков последовательностей $ x_i(k) $ и $
s_2(k) $ на интервале времени $ n \in [0, 5N - 1] $. Вычислить среднее значение, энергию и среднюю мощность последовательности $ s_2(k) $.

а. Операции при моделировании линейной комбинации сигналов $ s_2(k) $;

b. Как определяют указанные характеристики.

```python
k_long = np.arange(5 * N)  # n ∈ [0, 5N - 1]

# Моделирование отдельных гармонических сигналов
x1_k = B1 * np.sin(w1 * k_long)
x2_k = B2 * np.sin(w2 * k_long)
x3_k = B3 * np.sin(w3 * k_long)

# Линейная комбинация
s2_k = a1 * x1_k + a2 * x2_k + a3 * x3_k

# Визуализация отдельных сигналов и их комбинации
plt.figure(figsize=(15, 10))

# Сигнал x1(k)
plt.subplot(4, 1, 1)
plt.stem(k_long, x1_k, basefmt=" ", linefmt='C0-', markerfmt='C0o', label=f'$x_1(k) = {B1:.1f} \\cdot \\sin({w1:.3f}k)$')
plt.title('Гармонический сигнал $x_1(k)$')
plt.xlabel('Дискретное время k')
plt.ylabel('$x_1(k)$')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlim(0, 5*N)

# Сигнал x2(k)
plt.subplot(4, 1, 2)
plt.stem(k_long, x2_k, basefmt=" ", linefmt='C1-', markerfmt='C1s', label=f'$x_2(k) = {B2:.1f} \\cdot \\sin({w2:.3f}k)$')
plt.title('Гармонический сигнал $x_2(k)$')
plt.xlabel('Дискретное время k')
plt.ylabel('$x_2(k)$')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlim(0, 5*N)

# Сигнал x3(k)
plt.subplot(4, 1, 3)
plt.stem(k_long, x3_k, basefmt=" ", linefmt='C2-', markerfmt='C2^', label=f'$x_3(k) = {B3:.1f} \\cdot \\sin({w3:.3f}k)$')
plt.title('Гармонический сигнал $x_3(k)$')
plt.xlabel('Дискретное время k')
plt.ylabel('$x_3(k)$')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlim(0, 5*N)

# Линейная комбинация s2(k)
plt.subplot(4, 1, 4)
plt.stem(k_long, s2_k, basefmt=" ", linefmt='C3-', markerfmt='C3d')
plt.title(f'Линейная комбинация $s_2(k) = {a1:.1f}x_1(k) + {a2:.1f}x_2(k) + {a3:.1f}x_3(k)$')
plt.xlabel('Дискретное время k')
plt.ylabel('$s_2(k)$')
plt.grid(True, linestyle=':', alpha=0.7)
plt.xlim(0, 5*N)

plt.tight_layout()
plt.show()

# Вычисление характеристик сигнала s2(k)
mean_s2 = np.mean(s2_k)  # Среднее значение
E = np.sum(s2_k**2)      # Энергия сигнала
P = np.mean(s2_k**2)     # Средняя мощность

print("\n=== ХАРАКТЕРИСТИКИ СИГНАЛА s2(k) ===")
print(f"Среднее значение (mean_s2): {mean_s2:.6f}")
print(f"Энергия (E): {E:.6f}")
print(f"Средняя мощность (P): {P:.6f}")

# Визуализация результирующего сигнала
plt.figure(figsize=(12, 4))
plt.stem(k_long, s2_k, basefmt=" ", linefmt='C3-', markerfmt='C3d')
plt.title(f'Линейная комбинация гармонических сигналов $s_2(k)$')
plt.xlabel('Дискретное время k')
plt.ylabel('$s_2(k)$')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

### a. Операции при моделировании линейной комбинации сигналов $s_2(k)$

Моделирование линейной комбинации $s_2(k)$ основывается на **принципе суперпозиции** для линейных систем. Сигнал $s_2(k)$ является взвешенной суммой трех дискретных гармонических сигналов $x_1(k), x_2(k)$ и $x_3(k)$.

#### 1. Генерация гармонических компонентов $x_i(k)$:

Каждый базовый сигнал $x_i(k)$ моделируется как дискретная синусоида:

$$
x_i(k) = B_i \sin(\hat{\omega}_i k), \quad i = 1, 2, 3
$$

где:
- $k \in [0, 5N - 1]$ — дискретное время
- $B_i$ — амплитуда $i$-го компонента
- $\hat{\omega}_i$ — цифровая частота $i$-го компонента (в радианах на отсчет)

На этом этапе создаются три независимых последовательности, каждая из которых имеет свою амплитуду и частоту.

#### 2. Масштабирование (Умножение на коэффициент $a_i$):

Каждый компонент $x_i(k)$ умножается на соответствующий весовой коэффициент $a_i$:

$$
a_i x_i(k) = a_i \cdot B_i \sin(\hat{\omega}_i k)
$$

Эта операция определяет, насколько сильно $i$-й компонент будет влиять на итоговый сигнал $s_2(k)$.

#### 3. Сложение (Суперпозиция):

Конечный сигнал $s_2(k)$ получается путем поэлементного сложения всех взвешенных компонентов:

$$
s_2(k) = \sum_{i=1}^3 a_i x_i(k) = a_1 x_1(k) + a_2 x_2(k) + a_3 x_3(k)
$$

В результате $s_2(k)$ является составным квазипериодическим сигналом, содержащим все частоты $\hat{\omega}_1, \hat{\omega}_2, \hat{\omega}_3$.

### b. Как определяют указанные характеристики

Расчет характеристик производится для дискретной последовательности $s_2(k)$ на заданном интервале длины $L = 5N$.

#### 1. Среднее значение ($\overline{s_2(k)}$):

Среднее значение последовательности $s_2(k)$ определяется как среднее арифметическое всех отсчетов $s_2(k)$ на интервале $L$.

$$
\overline{s_2(k)} = \frac{1}{L} \sum_{k=0}^{L-1} s_2(k)
$$

#### 2. Энергия (E):

Энергия дискретной последовательности $s_2(k)$ определяется как сумма квадратов абсолютных значений ее отсчетов за весь интервал $L$. Поскольку сигнал вещественный, абсолютное значение опускается:

$$
E = \sum_{k=0}^{L-1} |s_2(k)|^2 = \sum_{k=0}^{L-1} s_2^2(k)
$$

#### 3. Средняя мощность (P):

Средняя мощность $P$ определяется как **энергия**, **деленная на длину последовательности** $L$. Это средняя энергия, приходящаяся на один отсчет, что делает эту характеристику независимой от длины интервала.

$$
P = \frac{1}{L} \sum_{k=0}^{L-1} |s_2(k)|^2
$$

## Задание 3

Вывести график дискретного сигнала $ s_3(k) $, представляющего собой дискретный гармонический сигнал $ s(k) $

$ s(k) = C \cos(\hat{\omega_0} k) $

с экспоненциальной огибающей $ |a|^k $, на интервале времени $ n \in [0, N - 1] $.

Привести аналитическую формулу дискретного сигнала $ s_3(k) $ и пояснить операции при его моделировании.

```python
# Вектор дискретного времени
k_3 = np.arange(N)

# Базовый гармонический сигнал
s_cos = C * np.cos(w0 * k_3)

# Экспоненциальная огибающая
exp_envelope = np.abs(a) ** k_3

# Результирующий сигнал с экспоненциальной огибающей
s3_k = s_cos * exp_envelope

# Визуализация сигнала
plt.figure(figsize=(12, 6))
plt.stem(k_3, s3_k, basefmt=" ", linefmt='C2-', markerfmt='C2o', label='$s_3(k)$')
plt.plot(k_3, exp_envelope, 'r--', linewidth=2, label=f'Верхняя огибающая $|{a:.3f}|^k$')
plt.plot(k_3, -exp_envelope, 'r--', linewidth=2, label=f'Нижняя огибающая $-|{a:.3f}|^k$')
plt.title(f'Сигнал $s_3(k)$ с экспоненциальными огибающими')
plt.xlabel('Дискретное время k')
plt.ylabel('$s_3(k)$')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

Аналитическая формула для дискретного сигнала $s_3(k)$ имеет вид:

$s_3​(k)$ =C ⋅ $∣a∣^k$ ⋅ cos($ω_0$k​) , k=0,1,…,N−1

При моделировании сигнала выполняется расчет экспоненциальной огибающей согласно текущему временному отсчету k, после расчитывается гармонический сигнал cos($ω_0$k​), итоговый сигнал $s_3​(k)$ формируется как произведение огибающей и гармонической составляющей.


## Задание 4

Вывести график пяти периодов периодической последовательности $ s_4(k) $ дискретных прямоугольных импульсов амплитуды $U$ и длительности $n_{imp}$ с периодом, вдвое большим длительности импульса.

Пояснить операции при моделировании периодической последовательности.

```python
# Период последовательности (вдвое больше длительности импульса)
T_period = 2 * n_imp
# Количество периодов для отображения
num_periods = 5
total_length = num_periods * T_period

# Создаем один период сигнала
one_period = np.zeros(T_period)
one_period[:n_imp] = U

# Создаем полную периодическую последовательность
k_4 = np.arange(total_length)
s4_k = np.tile(one_period, num_periods)

# Визуализация
plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
plt.stem(k_4, s4_k, basefmt=" ", linefmt='C0-', markerfmt='C0o')
plt.title(f'Периодическая последовательность прямоугольных импульсов $s_4(k)')
plt.xlabel('Дискретное время k')
plt.ylabel('$s_4(k)$')
plt.grid(True, linestyle=':', alpha=0.9)

# Разметка периодов
for i in range(num_periods + 1):
    plt.axvline(x=i * T_period, color='red', linestyle='--', alpha=0.5)
    if i < num_periods:
        plt.text(i * T_period + T_period/2, U + 0.5, f'Период {i+1}',
                ha='center', va='bottom', fontsize=10)
```

Аналитическая формула для периодической последовательности $s_4(k)$ имеет вид:
$$
s_4(k) = \begin{cases}
U, & если, (k  \mod  T) < n_{imp} \\
0, & иначе
\end{cases}
$$
При моделировании последовательности определяется период T = 2 ⋅ $n_{imp}$, определяется число временных отсчетов, как 5 T -1, в каждом периоде первые  $n_{imp}$ равны U, остальные нулю.
