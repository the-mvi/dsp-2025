## Основные теоретические положения


В лабораторной работе будут изучены следующие виды дискретных сигналов:

**Единичный цифровой импульс $δ_d(k)$:**
$$δ_d(k) = \begin{cases}
1, & k = 0  \\
0, & k \neq 0
 \end{cases}$$

**Дискретный единичный скачок $σ_d(k)$:**
$$σ_d(k) = \begin{cases}
1, & k \geq 0 \\
0, & k < 0
 \end{cases}$$

**Дискретная экспоненциальная функция $s(k)$:**
$$s(k) = \begin{cases}
a^k, & k \geq 0 \\
0, & k < 0
\end{cases}$$
**Дискретный комплексный гармонический сигнал $s(k)$:**
$$s(k) = Ce^{iw_0k}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Создание переменных в соответствии с вариантом
Nb = 2
N = 32
T = 0.0015
a = 0.81
C = 3
w0 = np.pi / 8
m = 7
```

1.Смоделировать единичный цифровой импульс $δ_d$(k) с выводом графиков на интервале дискретного времени nT ∈ [0, (N − 1)T] и дискретного нормированного времени n ∈ [0, N − 1].

Пояснить:
- взаимосвязь между дискретным и дискретным нормированным временем;
- различие между единичным цифровым импульсом и дельта-функцией.

```python
# Создание графика на интервале дискретного времени
t1 = np.linspace(0, (N-1) * T, N)
delta1 = np.zeros_like(t1)
delta1[0] = 1

# Создание графика на интервале дискретного нормального времени
t2 = np.linspace(0, N-1, N)
delta2 = np.zeros_like(t2)
delta2[0] = 1

# Вывод графиков
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.stem(t1, delta1, linefmt='grey', basefmt='grey')
plt.subplot(1, 2, 2)
plt.stem(t2, delta2, linefmt='grey', basefmt='grey')
```

- Дискретное время $-$ множество равноудаленных точек с периодом $T$, т.е $nT$, где $n \in Z$, а дискретное нормированное время получается при деление дискретного времени на период, т.е.  $n$, где $n \in Z$.

- Единичный цифровой импульс $-$ дискретный аналог дельта-функции:
$$δ(x) = \begin{cases}
∞, & x = 0 \\
0, & x \neq 0
\end{cases},$$
Принимающий в точке $0$ значение $1$  

Смоделировать дискретный единичный скачок $σ_d(k)$ с выводом графиков на интервале дискретного времени nT ∈ [0, (N − 1)T] и дискретного нормированного времени n ∈ [0, N − 1]. Пояснить: соответствие между дискретным и аналоговым единичными скачками; чему равна частота дискретизации дискретного единичного скачка.

```python
sigma1 = np.array(delta1)
sigma2 = np.array(delta2)

sigma1[:] = 1
sigma2[:] = 1

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.stem(t1, sigma1, linefmt='grey', basefmt='grey')
plt.subplot(1, 2, 2)
plt.stem(t2, sigma2, linefmt='grey', basefmt='grey')
```

- И дискретный и аналоговый еденичные скачки принимают значение $0$ при аргументе меньше $0$ и $1$ при аргументе больше $0$, но дискретный скачек является функцией от дискретного времени
- Частота дискретизации равна: $f_d = \frac{1}{T} ≈ 667$

Смоделировать дискретную экспоненциальную функцию $s_1(k)$ с выводом графиков на интервале дискретного времени $nT ∈ [0, (N − 1)T]$ и дискретного нормированного времени $n ∈ [0, N − 1]$. Пояснить соответствие между дискретной и аналоговой экспонентами.

```python
s11 = np.pow(a, t1)
s12 = np.pow(a, t2)

plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.stem(t1, s11, linefmt='grey', basefmt='grey')
plt.subplot(1, 2, 2)
plt.stem(t2, s12, linefmt='grey', basefmt='grey')
```

- Вид обоих функций определяется значением основания экспоненты $a$.

Смоделировать дискретный комплексный гармонический сигнал $s_2(k) = C \exp(j \hat ω_{0}k)$ с выводом графиков вещественной и мнимой частей на интервале времени n ∈ [0, N − 1]. Записать данный сигнал в виде комбинации двух вещественных последовательностей.

```python
s2 = C * np.exp(1j * w0 * (t2))

plt.figure(figsize=(15, 5))

plt.stem(t2, s2.imag, linefmt='grey', basefmt='grey', label='Imaginary')
plt.stem(t2, s2.real, linefmt='green', basefmt='grey', label='Real')
plt.legend()

print("Real: ", " ".join(list(map(str, np.round(s2.real, 2)))))
print("Imaginary: ", " ".join(list(map(str, np.round(s2.imag, 2)))))
```

$s_2 = 3e^{j \frac{π}{8}k} = 3 (\cos(\frac{π}{8}k)+ j\sin(\frac{π}{8}k))$, отсюда сигнал можно разделить на две вещественные части:

- Действительную: $3\cos(\frac{π}{8}k)$
- Мнимую: $3\sin(\frac{π}{8}k)$

Их комбинация и будет изначальным сигналом.

Вывести графики последовательностей $δ_d(k)$, $σ_d(k)$ и $s_1$(k), задержанных на m отсчетов, на интервале времени n ∈ [0, N − 1]. Записать формулы задержанных последовательностей.


```python
delta_delayed = np.zeros_like(t2)
delta_delayed[m] = 1

sigma_delayed = np.zeros_like(t2)
sigma_delayed[m:] = 1

exp_delayed = np.pow(a, t2 - m)
exp_delayed[:m] = 0

plt.figure(figsize=(22, 5))
plt.subplot(1, 3, 1)
plt.stem(t2, delta_delayed, linefmt='grey', basefmt='grey')

plt.subplot(1, 3, 2)
plt.stem(t2, sigma_delayed, linefmt='grey', basefmt='grey')

plt.subplot(1, 3, 3)
plt.stem(t2, exp_delayed, linefmt='grey', basefmt='grey')
```

Формулы задержанных последовательностей:

- $δ_d^*(k) = δ_d(k - m)$
- $\sigma_d^*(k) = \sigma_d(k - m)$
- $s_1^*(k) = s_1(k - m)$
