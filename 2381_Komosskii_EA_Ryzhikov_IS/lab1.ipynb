## Основные теоретические положения

### Оси времени

- **Дискретное время**: $ t = nT $, где $ n = 0, 1, 2, \ldots $ - номер отсчета, $ T $ - период дискретизации.
- **Дискретное нормированное время**: $ n = 0, 1, 2, \ldots $ - номер отсчета.

### Единичный цифровой импульс

$$ \delta_d(k) = \begin{cases} 1, & k = 0 \\ 0, & k \neq 0 \end{cases} $$

### Дельта-функция Дирака

$$ \delta(t) = \begin{cases} \infty, & t = 0 \\ 0, & t \neq 0 \end{cases}, \quad \text{и} \quad \int_{-\infty}^{\infty} \delta(t) dt = 1 $$

### Единичный скачок

$$ \sigma_d(k) = \begin{cases} 1, & k \geq 0 \\ 0, & k < 0 \end{cases}, k \in \mathbb{Z} $$

### Единичный аналоговый скачок

$$ \sigma(t) = \begin{cases} 1, & t \geq 0 \\ 0, & t < 0 \end{cases}, t \in \mathbb{R} $$

### Частота дискретизации

$$ f_d = \frac{1}{T} $$

### Дискретная экспонента

$$ s_1(k) = \begin{cases} a^k, & k \geq 0 \\ 0, & k < 0 \end{cases}, k \in \mathbb{Z} $$

### Аналоговая экспонента

$$ s_1(t) = \begin{cases} a^t, & t \geq 0 \\ 0, & t < 0 \end{cases}, t \in \mathbb{R} $$

### Дискретный комплексный гармонический сигнал

$$ s_2(k) = C \exp(j \hat{\omega_0} k) = C (\cos(\hat{\omega_0} k) + j \sin(\hat{\omega_0} k)), k \in \mathbb{Z} $$

### Аналоговый комплексный гармонический сигнал

$$ s_2(t) = C \exp(j \omega_0 t) = C (\cos(\omega_0 t) + j \sin(\omega_0 t)), t \in \mathbb{R} $$

### Задержки в дискретных сигналах

- Задержанный импульс: $$ \delta_d(k - m) = \begin{cases} 1, & k = m \\ 0, & k \neq m \end{cases}, k \in \mathbb{Z} $$
- Задержанный скачок: $$ \sigma_d(k - m) = \begin{cases} 1, & k \geq m \\ 0, & k < m \end{cases}, k \in \mathbb{Z} $$
- Задержанная экспонента: $$ s_1(k - m) = a^{k - m} = \begin{cases} a^{k - m}, & k \geq m \\ 0, & k < m \end{cases}, k \in \mathbb{Z} $$

```python
# Импорт необходимых библиотек
import math
import numpy as np
import matplotlib.pyplot as plt

```

```python
Nb = 6
"""Номер бригады"""

N = 30 + Nb % 5
"""Длина последовательности"""

T = 5e-4*(1 + Nb % 3) 
"""Период дискретизации"""

a = (-1) ** (Nb) * (0.8 + 5e-3 * (Nb % 20))
"""Основание экспоненты"""

C = 1 + Nb % 5
"""Амплитуда гармонического сигнала"""

w0 = math.pi / (6 + Nb % 5)
"""Частота гармонического сигнала"""

m = 5 + Nb % 5
"""Задержка"""

Nb, N, T, a, C, w0, m
```

### Задание 1


Смоделировать единичный цифровой импульс $ \delta_d(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить:

1) Взаимосвязь между дискретным и дискретным нормированным временем
2) Различие между единичным цифровым импульсом и дельта-функцией


```python
# Дискретное нормированное время
n = np.arange(N)
# Дискретное время
t = n * T

# Единичный цифровой импульс
delta_d = np.zeros(N)
delta_d[0] = 1

# График по дискретному времени
plt.figure(figsize=(10,4))
plt.stem(t, delta_d, basefmt=" ")
plt.title("Единичный цифровой импульс δ_d(k) во времени nT")
plt.xlabel("t, c")
plt.ylabel("δ_d(k)")
plt.grid(True)
plt.show()

# График по нормированному времени
plt.figure(figsize=(10,4))
plt.stem(n, delta_d, basefmt=" ")
plt.title("Единичный цифровой импульс δ_d(k) в нормированном времени n")
plt.xlabel("n")
plt.ylabel("δ_d(k)")
plt.grid(True)
plt.show()
```

Дискретное время $ t $ и дискретное нормированное время $ n $ связаны соотношением $ t = nT $, где $ T $ - период дискретизации.

Таким образом, дискретное нормированное время является безразмерной величиной, представляющей собой номер отсчета, в то время как дискретное время измеряется в секундах и зависит от выбранного периода дискретизации.

Единичный цифровой импульс 
$$ \delta_d(k) = \begin{cases} 1, & k = 0 \\ 0, & k \neq 0 \end{cases}, k \in \mathbb{Z} $$

В то время как дельта-функция Дирака $ \delta(t) $ является идеализированной функцией, которая не является функцией в традиционном смысле, а скорее распределением. Она определяется следующим образом:

$$ \delta(t) = \begin{cases} \infty, & t = 0 \\ 0, & t \neq 0 \end{cases}, \quad \text{и} \quad \int_{-\infty}^{\infty} \delta(t) dt = 1 $$

Единичный цифровой импульс применяется в дискретных системах, тогда как дельта-функция Дирака используется в непрерывных системах.


### Задание 2

Смоделировать дискретный единичный скачок $ \sigma_d(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить:
1) Соответствие между дискретным и аналоговым единичными скачками;
2) Чему равна частота дискретизации дискретного единичного скачка.


```python
# Дискретный единичный скачок σ_d(k): 1 для k >= 0, иначе 0
sigma_d = (n >= 0).astype(float)

# График в физическом времени t = nT
plt.figure(figsize=(10,4))
plt.stem(t, sigma_d, basefmt=" ")
plt.title(r"Дискретный единичный скачок $\sigma_d(k)$ во времени $t=nT$ (скачок при $n=0$)")
plt.xlabel("t, с")
plt.ylabel(r"$\sigma_d(k)$")
plt.grid(True)
plt.show()

# График в нормированном времени n
plt.figure(figsize=(10,4))
plt.stem(n, sigma_d, basefmt=" ")
plt.title(r"Дискретный единичный скачок $\sigma_d(k)$ в нормированном времени $n$ (скачок при $n=0$)")
plt.xlabel("n")
plt.ylabel(r"$\sigma_d(k)$")
plt.grid(True)
plt.show()
```

Единичный скачок в дискретном времени определяется как:

$$ \sigma_d(k) = \begin{cases} 1, & k \geq 0 \\ 0, & k < 0 \end{cases}, k \in \mathbb{Z} $$

Аналоговый единичный скачок определяется как:

$$ \sigma(t) = \begin{cases} 1, & t \geq 0 \\ 0, & t < 0 \end{cases}, t \in \mathbb{R} $$

Дискретный единичный скачок $ \sigma_d(k) $ является дискретной версией аналогового единичного скачка $ \sigma(t) $. Различие между ними заключается в том, что дискретный скачок определен только в дискретные моменты времени, тогда как аналоговый скачок определен для всех значений времени.

Частота дискретизации дискретного единичного скачка определяется как обратная величина периода дискретизации $ T $:
$$ f_d = \frac{1}{T} $$



```python
f_d = 1 / T
print(f"Период дискретизации T = {T} с")
print(f"Частота дискретизации f_d = {f_d} Гц")
```

### Задание 3

Смоделировать дискретную экспоненциальную функцию $ s_1(k) $ с выводом графиков на интервале дискретного времени $ nT \in [0, (N-1)T] $ и дискретного нормированного времени $ n \in [0, N-1] $. Пояснить соответствие между дискретной и аналоговой экспонентами.
 

```python
s1 = a**n

# График по физическому времени
plt.figure(figsize=(10,4))
plt.stem(t, s1, basefmt=" ")
plt.title(r"Дискретная экспоненциальная функция $s_1(k)$ во времени $t=nT$")
plt.xlabel("t, с")
plt.ylabel(r"$s_1(k)$")
plt.grid(True)
plt.show()

# График по нормированному времени
plt.figure(figsize=(10,4))
plt.stem(n, s1, basefmt=" ")
plt.title(r"Дискретная экспоненциальная функция $s_1(k)$ в нормированном времени $n$")
plt.xlabel("n")
plt.ylabel(r"$s_1(k)$")
plt.grid(True)
plt.show()
```

Дискретная экспоненциальная функция определяется как:

$$ s_1(k) = \begin{cases} a^k, & k \geq 0 \\ 0, & k < 0 \end{cases}, k \in \mathbb{Z} $$



Аналоговая экспоненциальная функция определяется как:

$$ s(t) = \begin{cases} a^t, & t \geq 0 \\ 0, & t < 0 \end{cases}, t \in \mathbb{R} $$

где $a$ — основание экспонент.

Дискретная экспоненциальная функция $ s_1(k) $ является дискретной версией аналоговой экспоненциальной функции $ s(t) $. Различие между ними заключается в том, что дискретная экспонента определена только в дискретные моменты времени, тогда как аналоговая экспонента определена для всех значений времени.

При этом, если подставить в аналоговую экспоненту дискретные значения времени $ t = kT $, то можно получить дискретную экспоненту с учетом периода дискретизации $ T $.

### Задание 4

Смоделировать дискретный комплексный гармонический сигнал $s_2(k) = C \exp(j \hat{\omega_0} k)$ с выводом графиков вещественной и мнимой частей на интервале времени $ n \in [0, N-1] $ Записать данный сигнал в виде комбинации двух вещественных последовательностей.

```python
# Комплексный сигнал
s2 = C * np.exp(1j * w0 * n)

# Вещественная и мнимая части
s2_real = np.real(s2)
s2_imag = np.imag(s2)

# Графики
plt.figure(figsize=(10,4))
plt.stem(n, s2_real, basefmt=" ", linefmt="b-", markerfmt="bo", label="Re{s2(k)}")
plt.stem(n, s2_imag, basefmt=" ", linefmt="r-", markerfmt="ro", label="Im{s2(k)}")
plt.title(r"Дискретный гармонический сигнал $s_2(k)$")
plt.xlabel("n")
plt.ylabel("Амплитуда")
plt.grid(True)
plt.legend()
plt.show()
```

Комплексный гармонический сигнал можно записать в виде:

$$ s_2(k) = C \exp(j \hat{\omega_0} k) = C (\cos(\hat{\omega_0} k) + j \sin(\hat{\omega_0} k)) $$

По формуле Эйлера, где $ C $ - амплитуда сигнала, $ \hat{\omega_0} $ - нормированная частота в радианах на отсчет, $ j $ - мнимая единица.

Таким образом, можно выделить вещественную и мнимую части сигнала:

- Вещественная часть: $ \text{Re}(s_2(k)) = C \cos(\hat{\omega_0} k) $
- Мнимая часть: $ \text{Im}(s_2(k)) = C \sin(\hat{\omega_0} k) $
  

### Задание 5

Вывести графики последовательностей  $\delta_d(k), \sigma_d(k), \text{и } s_1(k)$ задержанных на $m $ отсчетов, на интервале времени $n \in [0, N-1]$ Записать формулы задержанных последовательностей.

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
Nb = 6
N = 30 + Nb % 5
m = 5 + Nb % 5
a = (-1)**Nb * (0.8 + 0.005 * (Nb % 20))

# Ось дискретного времени
n = np.arange(N)

# Задержанный импульс
delta_d = np.zeros(N)
delta_d[m] = 1

# Задержанный скачок
sigma_d = (n >= m).astype(float)

# Задержанная экспонента
s1 = np.zeros(N)
s1[n >= m] = a**(n[n >= m] - m)

# Построение графиков
plt.figure(figsize=(12,8))

plt.subplot(3,1,1)
plt.stem(n, delta_d, basefmt=" ")
plt.title(r"Задержанный единичный импульс $\delta_d(k-m)$")
plt.xlabel("n")
plt.ylabel("Амплитуда")
plt.grid(True)

plt.subplot(3,1,2)
plt.stem(n, sigma_d, basefmt=" ")
plt.title(r"Задержанный единичный скачок $\sigma_d(k-m)$")
plt.xlabel("n")
plt.ylabel("Амплитуда")
plt.grid(True)

plt.subplot(3,1,3)
plt.stem(n, s1, basefmt=" ")
plt.title(r"Задержанная дискретная экспонента $s_1(k-m)$")
plt.xlabel("n")
plt.ylabel("Амплитуда")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Формулы задержанных последовательностей:

- Задержанный импульс:
$$ \delta_d(k - m) = \begin{cases} 1, & k = m \\ 0, & k \neq m \end{cases}, k \in \mathbb{Z} $$
- Задержанный скачок:
$$ \sigma_d(k - m) = \begin{cases} 1, & k \geq m \\ 0, & k < m \end{cases}, k \in \mathbb{Z} $$
- Задержанная экспонента:
$$ s_1(k - m) = a^{k - m} = \begin{cases} a^{k - m}, & k \geq m \\ 0, & k < m \end{cases}, k \in \mathbb{Z} $$
