Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. #### Вывод передаточной функции для формулы прямоугольников

Формула прямоугольников: $y_{n+1} = y_n + s_{n+\frac{1}{2}}$

Пусть входной сигнал $s_n = e^{i \omega n}$, тогда выходной сигнал будет равен $y_n = H(\omega) e^{i \omega n}$, где $H(\omega)$ - передаточная функция фильтра.

Подставим входной и выходной сигналы в формулу прямоугольников:

$$H(\omega) e^{i \omega (n+1)} = H(\omega) e^{i \omega n} + e^{i \omega (n+\frac{1}{2})}$$

Выразим $H(\omega)$:

$$
H(\omega) =
\frac{e^{i \omega / 2}}{e^{i \omega} - 1} =
\frac{1}{e^{i \omega / 2} - e^{-i \omega / 2}} =
\frac{1}{2i \sin(\omega / 2)}
$$

Итого, передаточная функция для формулы прямоугольников равна:

$$
\tilde{H}(f) = H(2 \pi f) =
\frac{1}{2i \sin(\pi f)}
$$


Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i\omega}$

Вычисленное значение интеграла с помощью формулы прямоугольников равно $H(\omega) e^{i \omega t} = \frac{e^{i \omega t}}{2i \sin(\omega / 2)}$

Отношение вычисляемого в результате фильтрации значения к истинному равно:

$$
\gamma = \frac{Вычисленное\ значение}{Истинное\ значение} =
\frac{\omega /2}{\sin(\omega / 2)}
$$


___

#### Вывод передаточной функции для формулы прямоугольников

Формула трапеций: $y_{n+1} = y_n + \frac{1}{2}(s_n + s_{n+1})$

Подставим входной и выходной сигналы в формулу трапеций:

$$H(\omega)e^{i\omega(n+1)} = H(\omega)e^{i\omega n} + \frac{1}{2}(e^{i\omega n} + e^{i\omega(n+1)})$$

Выразим $H(\omega)$:

$$H(\omega) = 
\frac{1 + e^{i\omega}}{2(e^{i\omega} - 1)} =
\frac{\cos(\frac{\omega}{2})}{2i\sin(\frac{\omega}{2})}$$

Итого, передаточная функция для формулы трапеций равна:

$$
\tilde{H}(f) = \frac{\cos(\pi f)}{2i\sin(\pi f)}
$$

Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i\omega}$

Вычисленное значение интеграла с помощью формулы трапеций равно $H(\omega) e^{i \omega t} = \frac{\cos(\frac{\omega}{2})e^{i \omega t}}{2i\sin(\frac{\omega}{2})}$

Отношение вычисляемого в результате фильтрации значения к истинному равно:

$$
\gamma = \frac{Вычисленное значение}{Истинное значение} = \frac{\omega /2\cos(\frac{\omega}{2})}{\sin(\frac{\omega}{2})}
$$

___

#### Вывод передаточной функции для формулы Симпсона

Формула Симпсона: $y_{n+1} = y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1})$

Пусть входной сигнал $s_n = e^{i \omega n}$, тогда выходной сигнал будет равен $y_n = H(\omega) e^{i \omega n}$, где $H(\omega)$ - передаточная функция фильтра.

Подставим входной и выходной сигналы в формулу Симпсона:

$$H(\omega)e^{i\omega(n+1)} = H(\omega)e^{i\omega(n-1)} + \frac{1}{3}(e^{i\omega(n-1)} + 4e^{i\omega n} + e^{i\omega(n+1)})$$

Выразим $H(\omega)$:

$$
H(\omega) =
\frac{e^{i\omega} + 4 + e^{-i\omega}}{3(e^{i\omega} - e^{-i\omega})} = \frac{4+ 2\cos(\omega)}{6i\sin(\omega)} = \frac{2 + \cos(\omega)}{3i\sin(\omega)}
$$

Итого, передаточная функция для формулы Симпсона равна:

$$
\tilde{H}(f) = H(2 \pi f)
= \frac{2 + \cos(2 \pi f)}{3i\sin(2 \pi f)}
$$


Точное значение интеграла функции $s(t) = e^{i \omega t}$ по-прежнему равно $\frac{e^{i \omega t}}{i\omega}$.

Вычисленное значение интеграла с помощью формулы Симпсона равно $H(\omega) e^{i \omega t} = \frac{2 + \cos(\omega)}{3i\sin(\omega)} e^{i \omega t}$

Отношение вычисляемого в результате фильтрации значения к истинному равно:

$$
\gamma = \frac{Вычисленное\ значение}{Истинное\ значение} =
\frac{2 + \cos(\omega)}{3} \cdot \frac{\omega}{\sin(\omega)}
$$


```python
f = np.linspace(1e-8, 0.5, 300, dtype=np.float64)

# Передаточная функция для формулы прямоугольников
H_rect = 1 / (2j * np.sin(np.pi * f))
# Передаточная функция для формулы трапеций
H_trap = np.cos(np.pi * f) / (2j * np.sin(np.pi * f))
# Передаточная функция для формулы Симпсона
H_simp = (2 + np.cos(2 * np.pi * f)) / (3j * np.sin(2 * np.pi * f))

imag_rect = np.imag(H_rect)
imag_trap = np.imag(H_trap)
imag_simp = np.imag(H_simp)

imag_rect[abs(imag_rect) > 15] = np.nan
imag_trap[abs(imag_trap) > 15] = np.nan
imag_simp[abs(imag_simp) > 15] = np.nan


plt.figure(figsize=(10, 6))
plt.plot(f, imag_rect, label='Формула прямоугольников', color='blue')
plt.plot(f, imag_trap, label='Формула трапеций', color='orange')
plt.plot(f, imag_simp, label='Формула Симпсона', color='green')

plt.title("Комплексная часть передаточных функций")
plt.xlabel("f")
plt.ylabel("Img H(f)")
plt.legend()
plt.grid()

plt.show()
```

```python
f = np.linspace(1e-8, 0.5, 300)
omega = 2 * np.pi * f

# Отношение вычисляемого значения к истинному для формулы прямоугольников
gamma_rect = (omega / 2) / np.sin(omega / 2)
# Отношение вычисляемого значения к истинному для формулы трапеций
gamma_trap = np.cos(omega / 2) * (omega / 2) / np.sin(omega / 2)
# Отношение вычисляемого значения к истинному для формулы Симпсона
gamma_simp = (2 + np.cos(omega)) / 3 * (omega) / np.sin(omega)




plt.figure(figsize=(10, 6))
plt.plot(f, gamma_rect, label='Формула прямоугольников', color='blue')
plt.plot(f, gamma_trap, label='Формула трапеций', color='orange')
plt.plot(f, gamma_simp, label='Формула Симпсона', color='green')

plt.title("Отношение вычисляемого значения к истинному")
plt.xlabel("f")
plt.ylabel("γ")
plt.legend()
plt.grid()

plt.ylim(0, 2)
plt.show()
```

#### Частотные характеристики передаточных функций

По полученным графиками передаточных функций и графиками отношения вычисляемого в результате фильтрации значения к истинному можно сделать следующие выводы:

- Общая характеристика всех трёх фильтров заключается в усилении низких частот — в этом диапазоне передаточные функции стремятся к бесконечности, а отношение вычисляемого значения к истинному стремится к единице. Это говорит о том, что все три метода хорошо справляются с интегрированием низкочастотных сигналов, так как идеальный интегратор имеет передаточную функцию $H(\omega) = \frac{1}{i\omega} \rightarrow -i / \omega$.
- Фильтр на основе формулы прямоугольников на высоких частотах сильно искажает сигнал, так как передаточная функция близка к нулю, а отношение вычисляемого значения к истинному больше единицы, что говорит о переоценке амплитуды высокочастотных сигналов.
- Фильтр на основе формулы трапеций имеет плавное затухание на высоких частотах — подавляет высокочастотные шумы лучше, чем формула прямоугольников.
- Фильтр на основе формулы Симпсона показывает более точные значения на низких частотах, но на высоких частотах значительно усиливает сигнал, что приводит к сильным искажениям.


### Задание 2

Вывести формулу передаточной функции рекурсивного фильтра для интегрирования по «правилу 3/8»:

$$ y_{n + 2} = y_{n - 1} + \frac{1}{8} (s_{n + 2} + 3s_{n + 1} + 3s_n + s_{n - 1}) $$

Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.

Пусть входной сигнал $s_n = e^{i \omega n}$, тогда выходной сигнал будет равен $y_n = H(\omega) e^{i \omega n}$, где $H(\omega)$ - передаточная функция фильтра.

Подставим входной и выходной сигналы в формулу «правила 3/8»:

$$H(\omega)e^{i\omega(n+2)} = H(\omega)e^{i\omega (n - 1)} + \frac{1}{8}(e^{i\omega (n + 2)} + 3 e^{i\omega(n+1)} + 3 e^{i \omega n} + e^{i \omega (n - 1)})$$

Выразим $H(\omega)$:

$$
H(\omega) =
\frac{e^{\frac{3i}{2} \omega} + 3 e^{\frac{i}{2} \omega} + 3 e^{-\frac{i}{2} \omega} + e^{-\frac{3i}{2} \omega}}{8(e^{\frac{3i}{2} \omega} - e^{-\frac{3i}{2} \omega})} =
\frac{\cos(\frac{3\omega}{2}) + 3\cos(\frac{\omega}{2})}{8i \sin(\frac{3 \omega}{2})}
$$

Итого, передаточная функция для формулы прямоугольников равна:

$$
\tilde{H}(f) = \frac{\cos(3 \pi f) + 3\cos(\pi f)}{8i \sin(3 \pi f)}
$$


Точное значение интеграла $e^{i \omega t}$ равно $\frac{e^{i \omega t}}{i\omega}$

Вычисленное значение интеграла при помощи формулы «правила 3/8» равно $H(\omega) e^{i \omega t} = \frac{\cos(\frac{3\omega}{2}) + 3\cos(\frac{\omega}{2})}{8i \sin(\frac{3 \omega}{2})} e^{i \omega t}$

Отношение вычисляемого в результате фильтрации значения к истинному равно:
$$
\gamma = \frac{Вычисленное\ значение}{Истинное\ значение} =
\frac{ \left( \cos(\frac{3\omega}{2}) + 3\cos(\frac{\omega}{2}) \right) \cdot \omega }{ 8 \sin(\frac{3\omega}{2}) }
$$

```python
f = np.linspace(1e-8, 0.5, 1000)

# Передаточная функция для правила 3/8
H_three_eight = (np.cos(3 * np.pi * f) + 3 * np.cos(np.pi * f)) / (
    8j * np.sin(3 * np.pi * f)
)

imag_three_eight = np.imag(H_three_eight)
imag_three_eight[abs(imag_three_eight) > 15] = np.nan

plt.figure(figsize=(10, 6))
plt.plot(f, imag_three_eight, label='Правило 3/8', color='purple')
plt.axvline(x=1/3, color='purple', linestyle='--', label='f = 1/3 (особая точка)', alpha=0.7)
plt.title("Комплексная часть передаточной функции для правила 3/8")
plt.xlabel("f")
plt.ylabel("Img H(f)")
plt.legend()
plt.grid()
plt.show()
```

```python
f = np.linspace(1e-8, 0.5, 3000)

omega = 2 * np.pi * f

# Отношение вычисляемого значения к истинному для правила 3/8
gamma_three_eight = (
    (np.cos(3 * omega / 2) + 3 * np.cos(omega / 2))
    * omega
    / (8 * np.sin(3 * omega / 2))
    
)

# gamma_three_eight[abs(gamma_three_eight) > 3] = np.nan

plt.figure(figsize=(10, 6))
plt.plot(f, gamma_three_eight, label='Правило 3/8', color='purple')
plt.title("Отношение вычисляемого значения к истинному для правила 3/8")
plt.xlabel("f")
plt.ylabel("γ")
plt.legend()
plt.grid()
plt.ylim(-3, 3)
plt.show()
```

Формула 3/8 обеспечивает высокую точность интегрирования на низких частотах, но обладает внутренним резонансом при
$f = \frac{1}{3}$, где передаточная функция обращается в бесконечность — метод теряет устойчивость.

Также $\gamma(0) = 1/3$ в данной формуле, когда в остальных $\gamma(0) = 1$.


### Задание 3

Провести сопоставительный анализ частотных характеристик передаточных функций, полученных при выполнении п. 1 и 2.

Все рассматриваемые методы реализуют численный интегратор, который усиливает низкие частоты и подавляет высокие. Их передаточные функции ведут себя как $H(\omega) \sim \frac{1}{i\omega}$ при $\omega \to 0$, что соответствует идеальному интегратору. Отношение $\gamma$ стремится к единице (кроме метода 3/8) при низких частотах для всех методов, что говорит о высокой точности интегрирования в этом диапазоне.

Однако, на высоких частотах методы существенно различаются по своим характеристикам.

1. Формула прямоугольников:
   - На низких частотах: хорошее совпадение с идеальным интегратором ($\gamma(0)=1$).
   - На высоких частотах: модуль $H(f)$ не падает достаточно быстро и $\gamma(f)>1$, то есть высокочастотная составляющая **переинтегрируется (амплитуда завышается)** и появляется заметная фазовая ошибка.
   - Вывод: пригоден только для узкой низкочастотной полосы.
2. Формула трапеций:
   - На низких частотах: $\gamma(0)=1$, аппроксимация близка к идеальному интегратору.
   - Вплоть до окрестности половины частоты дискретизации поведение наиболее «ровное»: умеренное подавление высоких частот без резонансов и взлётов.
   - Вывод: лучший компромисс точность/устойчивость; хороший выбор для практического интегратора.
3. Формула Симпсона
   - Низкие частоты: высокая точность ($\gamma(f)$ дольше остаётся близкой к 1, чем у прямоугольников).
   - При приближении к верхней части спектра появляются сильные искажения и резкий рост усиления — возможны «пики» и неустойчивое поведение.
   - Вывод: очень точна в узкой НЧ-области, но чувствительна к высокочастотным компонентам → требует ограничения полосы.
4. Правило 3/8
   - Частотная характеристика имеет **внутренний полюс при (f = 1/3)**: в этой точке $|H(f)| \to \infty$, что означает внутренний резонанс и потерю устойчивости схемы как интегратора.
   - Для $\omega \to 0$ получается $\gamma(0) = 1/3$, т.е. интегратор по формуле 3/8 в заданной реализации **занижает** идеальное значение в 3 раза на постоянной составляющей. Это следствие выбранной нормировки коэффициента (1/8); умножение передаточной функции на 3 вернёт $\gamma(0)$ к 1.
   - Вывод: правило 3/8 в таком виде не является универсальным цифровым интегратором: оно очень точное по форме на низких частотах, но с неверным DC-масштабом и жёстким ограничением по полосе из-за резонанса при (f=1/3).
