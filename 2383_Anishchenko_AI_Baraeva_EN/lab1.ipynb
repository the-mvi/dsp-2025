## Основные теоретические положения

Частота дискретизации:

$ f_d = \frac{1}{T} $

Единичный цифровой импульс:

$ \delta_d(k) =
\begin{cases}
1, & k = 0, \\
0, & k \neq 0,
\end{cases}
\quad k \in \mathbb{Z} $

Дельта-функция (функция Дирака):

$ \delta(t) =
\begin{cases}
\infty, & t = 0, \\
0, & t \neq 0.
\end{cases} $

Дискретный единичный скачок:

$ \sigma_d(k) =
\begin{cases}
1, & k \geq 0, \\
0, & k < 0,
\end{cases}
\quad k \in \mathbb{Z} $

Функция единичного скачка (функция Хевисайда):

$ \sigma(t) =
\begin{cases}
1, & t > 0, \\
0.5 \ или \ не \ определена, & t = 0, \\
0, & t < 0.
\end{cases} $

Дискретная экспоненциальная функция:

$ s_1(k) =
\begin{cases}
0, & k < 0, \\
a^k, & k \geq 0,
\end{cases}
\quad k \in \mathbb{Z} $

Дискретный комплексный гармонический сигнал:

$ s_2(k) = C \exp\!\big(j \omega_0 k \big) $

## Выполнение работы

```python
import math as m
import cmath as cm
import numpy as np
import matplotlib.pyplot as plt


Nb = 12
N = 30 + Nb % 5
T = 0.0005 * (1 + Nb % 3)
a = (-1) ** Nb * (0.8 + 0.005 * (Nb % 20))
C = 1 + Nb % 5
w0 = m.pi / (6 + Nb % 5)
m = 5 + Nb % 5
print(f'Nb = {Nb}, N = {N}, T = {T}, a = {a}, C = {C}, w0 = {w0}, m = {m}')
```

```python
# Функция единичного цифрового импульса
def delta_d(k):
  return 1 if k == 0 else 0

# Функция дискретного единичного скачка
def sigma_d(k):
  return 1 if k >= 0 else 0

# Дискретная экспоненциальная функция
def s1(k):
  return 0 if k < 0 else a ** k

# Дискретная экспоненциальная функция
def s2(k):
  return C * cm.exp(1j * w0 * k)

# Функция рисования графика
def graph_build(x, y, title, xlabel, ylable):
  plt.figure()
  plt.stem(x, y, markerfmt=".")
  plt.title(title)
  plt.xlabel(xlabel)
  plt.ylabel(ylable)
  plt.show()
```

```python
# Интервалы времени
n = np.arange(N)
nT = n * T
```

## Задание 1

```python
# Значения функции
Y = [delta_d(k) for k in n]
YT = [delta_d(k) for k in nT]

# График по нормированному времени
graph_build(n, Y, 'Единичный цифровой импульс', 'n', '$\\delta_d(n)$')

# График по дискретному времени
graph_build(nT, YT, 'Единичный цифровой импульс', 'nT, с', '$\\delta_d(nT)$')
```

a) Дискретное время и дискретное нормированное время связаны следующим образом:
$t = nT$, где $t$ — время в секундах, $n \in Z$, $T$ — шаг дискретизации.

b) Единичный цифровой импульс $ \delta_d(k) $ определён только для целых $ k $ и только в нуле его значение равно 1. Дельта-функция Дирака $ \delta(t) $ определена на множестве вещественных чисел и обращается в бесконечность только в нуле.

## Задание 2

```python
# Значения функции
Y = [sigma_d(k) for k in n]
YT = [sigma_d(k) for k in nT]

# График по нормированному времени
graph_build(n, Y, 'Дискретный единичный скачок', 'n', '$\\sigma_d(n)$')

# График по дискретному времени
graph_build(nT, YT, 'Дискретный единичный скачок', 'nT, с', '$\\sigma_d(nT)$')
```

a) Дискретный единичный скачок — это дискретное отображение аналогового единичного скачка (функции Хевисайда), примененное к дискретным моментам времени.

b) Частота дискретизации дискретного единичного скачка: $ f_d = \frac{1}{T} = \frac{1}{0.0005} = 2000 $ Гц

## Задание 3

```python
# Значения функции
Y = [s1(k) for k in n]
YT = [s1(k) for k in n]

# График по нормированному времени
graph_build(n, Y, 'Дискретная экспоненциальная функция', 'n', '$s_1(n)$')

# График по дискретному времени
graph_build(nT, YT, 'Дискретная экспоненциальная функция', 'nT, с', '$s_1(nT)$')
```

Дискретная экспонента — это последовательность значений непрерывной экспоненты, взятых с интервалом дискретизации во времени.

## Задание 4

```python
# Значения функции
Y_real = [np.real(s2(k)) for k in n]
Y_imag = [np.imag(s2(k)) for k in n]

# График по нормированному времени
graph_build(n, Y_real, 'Вещественная часть дискретного комплексного гармонического сигнала', 'n', '$s_2(n)$')

# График по дискретному времени
graph_build(n, Y_imag, 'Мнимая часть дискретного комплексного гармонического сигнала', 'n', '$s_2(n)$')
```

Сигнал записанный в виде комбинации двух вещественных последовательностей:
$ s_2(k) = C e^{j w_0 k} = C cos(w_0 k) + j \cdot C sin(w_0 k) = 3 cos(0.3927 k) + j \cdot 3 sin(0.3927 k) $

## Задание 5

```python
# Значения функций
Y_delta = [delta_d(k - m) for k in n]
Y_sigma = [sigma_d(k - m) for k in n]
Y_s1 = [s1(k - m) for k in n]

# Графики задержанных последовательностей
graph_build(n, Y_delta, 'Единичный цифровой импульс с задержкой', 'n', '$\\delta_d(n - m)$')
graph_build(n, Y_sigma, 'Дискретный единичный скачок с задержкой', 'n', '$\\sigma_d(n - m)$')
graph_build(n, Y_s1, 'Дискретная экспоненциальная функция с задержкой', 'n', '$s_1(n - m)$')
```

Формула единичного цифрового импульса с задержкой на $ m $ отсчетов:

$ \delta_d(k) =
\begin{cases}
1, & k = m, \\
0, & k \neq m,
\end{cases}
\quad k \in \mathbb{Z} $

Формула дискретного единичного скачка с задержкой на $ m $ отсчетов:

$ \sigma_d(k) =
\begin{cases}
1, & k \geq m, \\
0, & k < m,
\end{cases}
\quad k \in \mathbb{Z} $

Формула дискретной экспоненциальной функции с задержкой на $ m $ отсчетов:

$ s_1(k) =
\begin{cases}
0, & k < m, \\
a^{\,k-m}, & k \geq m,
\end{cases}
\quad k \in \mathbb{Z} $
