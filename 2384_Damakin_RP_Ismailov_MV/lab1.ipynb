## Основные теоретические положения

**Частота дискретизации**

$$
f_d = \frac{1}{T}
$$

Частота дискретизации — это количество выборок (отсчётов) непрерывного сигнала, которые снимаются за одну секунду. Она равна величине, обратной периоду дискретизации $T$. Чем выше $f_d$, тем точнее цифровая модель передаёт аналоговый сигнал, но тем больше данных нужно хранить и обрабатывать.



**Единичный цифровой импульс**

$$
\delta_d(k) =
\begin{cases}
1, & k = 0, \\[6pt]
0, & k \neq 0,
\end{cases}
\quad k \in \mathbb{Z}
$$

Это дискретная функция, принимающая значение 1 только в нулевой момент времени, а во всех остальных точках равная 0.



**Дельта-функция (функция Дирака)**

$$
\delta(t) =
\begin{cases}
\infty, & t = 0, \\[6pt]
0, & t \neq 0,
\end{cases}
$$

Дельта-функция — это идеализированная математическая модель мгновенного импульса бесконечной высоты и бесконечно малой ширины. Главное свойство: интеграл дельта-функции равен 1.



**Дискретный единичный скачок**

$$
\sigma_d(k) =
\begin{cases}
1, & k \geq 0, \\[6pt]
0, & k < 0,
\end{cases}
\quad k \in \mathbb{Z}
$$

Это дискретная функция, равная нулю до момента $k=0$, и единице начиная с этого момента.



**Функция единичного скачка (функция Хевисайда)**

$$
\sigma(t) =
\begin{cases}
1, & t > 0, \\[6pt]
0.5 \text{ или не определена}, & t = 0, \\[6pt]
0, & t < 0,
\end{cases}
$$

Аналог дискретного скачка в непрерывной области. Показывает резкое включение сигнала в момент времени $t=0$.



**Дискретная экспоненциальная функция**

$$
s_1(k) =
\begin{cases}
0, & k < 0, \\[6pt]
a^k, & k \geq 0,
\end{cases}
\quad k \in \mathbb{Z}
$$

Эта функция растёт (если $|a| > 1$) или затухает (если $|a| < 1$) экспоненциально после включения в момент $k=0$.



**Дискретный комплексный гармонический сигнал**

$$
s_2(k) = C \exp\!\big(j \omega_0 k \big)
$$

Это цифровая версия гармонического колебания. Здесь $C$ — амплитуда, $\omega_0$ — дискретная круговая частота, $j$ — мнимая единица.

## Выполнение работы

```python
import numpy as np
import cmath as cm
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nb = 15
N = 30 + Nb % 5
T = 0.0005 * (1 + Nb % 3)
a = (-1)**Nb * (0.8 + 0.005 * (Nb % 20))
C = 1 + Nb % 5
omega0_hat = np.pi / (6 + Nb % 5)
m = 5 + Nb % 5


print(f"Номер бригады: {Nb}")
print(f"Длина последовательности: {N}")
print(f"Период дискретизации: {T}")
print(f"Основание экспоненты: {a}")
print(f"Амплитуда гармонического сигнала: {C}")
print(f"Частота гармонического сигнала: {omega0_hat}")
print(f"Задержка: {m}")
```

```python
# Функция единичного цифрового импульса
def delta_d(k):
  return 1 if k == 0 else 0

# Функция дискретного единичного скачка
def sigma_d(k):
  return 1 if k >= 0 else 0

# Дискретная экспоненциальная функция
def s1(k):
  return 0 if k < 0 else a ** k

# Дискретный комплексный гармонический сигнал
def s2(k):
  return C * cm.exp(1j * omega0_hat * k)

# Функция для отрисовки графика
def graphic(x, y, title, xlabel, ylable):
  plt.figure(figsize=(8,4))
  plt.stem(x, y, basefmt="", markerfmt="o")
  plt.title(title)
  plt.xlabel(xlabel)
  plt.ylabel(ylable)
  plt.grid(True)
  plt.show()

n = np.arange(N)
nT = n * T
```

### **1.**

```python
Yn = np.array([delta_d(k) for k in n])
YnT = np.array([delta_d(k) for k in nT])


graphic(
    x=n,
    y=Yn,
    title="Единичный цифровой импульс $\\delta_d(n)$  в нормированном времени",
    xlabel="n",
    ylable="$\\delta_d(n)$"
)


graphic(
    x=nT,
    y=YnT,
    title="Единичный цифровой импульс $\\delta_d(n)$  во временной области",
    xlabel="t = nT",
    ylable="$\\delta_d(n)$"
)
```

**Пояснения:**

1. Дискретное время $nT$ задаёт отсчёты в физических единицах (секундах) и зависит от периода дискретизации $T$. Нормированное время $n$ отражает номер отсчёта без единиц измерения. Таким образом, форма импульса одинакова, но различается масштаб по оси времени.

2. Единичный цифровой импульс $\delta_d(k)$ является дискретной последовательностью (равен $1$ при $k=0$ и $0$ при $k \neq 0$), а дельта-функция $\delta(t)$ — это непрерывная обобщённая функция с интегралом, равным единице.


### **2.**

```python
Yn = np.array([sigma_d(k) for k in n])
YnT = np.array([sigma_d(k) for k in nT])

graphic(
    n,
    Yn,
    'Дискретный единичный скачок',
    'n',
    '$\\sigma_d(n)$'
  )

graphic(
    nT,
    YnT,
    'Дискретный единичный скачок',
    'nT, с',
    '$\\sigma_d(nT)$'
  )
```

**Пояснения:**

1. *Соответствие аналогового и дискретного скачков.*  
Дискретный единичный скачок $\sigma_d(n)$ представляет собой дискретизированный вариант функции Хэвисайда $\sigma(t)$ и задаётся соотношением $\sigma_d(n) = \sigma(nT)$.  
Главное отличие состоит в значении при $t=0$: для аналогового скачка оно либо не определено, либо принимается равным $\tfrac12$. В дискретном же случае обычно фиксируется значение $1$.

2. *Частота дискретизации.*  
Частота дискретизации определяется как $f_s = \tfrac{1}{T}$.  
Для $T = 0.0005 \, \text{с}$ получаем:  
$f_s = \tfrac{1}{0.0005} = 2000 \, \text{Гц}$.  
Следовательно, сигнал дискретизируется с частотой 2000 отсчётов в секунду.


### **3.**

```python
Yn = np.array([s1(k) for k in n])
YnT = np.array([s1(k) for k in n])

graphic(
    n,
    Yn,
    'Дискретная экспоненциальная функция',
    'n',
    '$s_1(n)$'
  )

graphic(
    nT,
    YnT,
    'Дискретная экспоненциальная функция',
    'nT, с',
    '$s_1(nT)$'
  )
```

**Пояснения:**

*Соответствие аналоговой и дискретной экспоненты.*

Дискретная экспоненциальная функция $s_1(n)$ является выборкой аналоговой экспоненты $s(t)$ по моментам времени $t = nT$, то есть $s_1(n) = s(nT)$.  
Дискретная экспонента повторяет форму аналоговой, но описана не как непрерывная кривая, а как набор отдельных значений в дискретные моменты времени. Характер затухания сохраняется.


### **4.**

```python
Yn_real = np.array([np.real(s2(k)) for k in n])
Yn_imag = np.array([np.imag(s2(k)) for k in n])

graphic(
    n,
    Yn_real,
    'Вещественная часть дискретного комплексного гармонического сигнала',
    'n',
    '$s_2(n)$'
  )

graphic(
    n,
    Yn_imag,
    'Мнимая часть дискретного комплексного гармонического сигнала',
    'n',
    '$s_2(n)$'
  )
```

**Пояснения:**

Формула Эйлера: $e^{j\alpha}=\cos\alpha + j\sin\alpha$.

Если применить к полученному сигналу:   
$ s_2(k) = C e^{j\omega_0 k} = C\cos(\omega_0 k) + j\,C\sin(\omega_0 k) = \cos\left(\tfrac{\pi}{6}k\right) + j\sin\left(\tfrac{\pi}{6}k\right) $


### **5.**

```python
Yn_delta = np.array([delta_d(k - m) for k in n])
Yn_sigma = np.array([sigma_d(k - m) for k in n])
Yn_s1 = np.array([s1(k - m) for k in n])

graphic(
    n,
    Yn_delta,
    'Единичный цифровой импульс с задержкой',
    'n',
    '$\\delta_d(n - m)$'
  )
graphic(
    n,
    Yn_sigma,
    'Дискретный единичный скачок с задержкой',
    'n',
    '$\\sigma_d(n - m)$'
  )
graphic(
    n,
    Yn_s1,
    'Дискретная экспоненциальная функция с задержкой',
    'n',
    '$s_1(n - m)$'
  )
```

**Формулы задержанных последовательностей:**

1. *Задержанный единичный цифровой импульс*  
   $\displaystyle \delta_d(n-m)=\delta[n-m]=\begin{cases}
     1, & n=m,\\[4pt]
     0, & n\neq m.
   \end{cases}$

2. *Задержанный единичный скачок*  
   $\displaystyle \sigma_d(n-m)=\begin{cases}
     1, & n\ge m,\\[4pt]
     0, & n<m.
   \end{cases}$

3. *Задержанная дискретная экспонента*  
   $\displaystyle s_1(n-m)=\begin{cases}
     b^{\,n-m}, & n\ge m,\\[4pt]
     0, & n<m.
   \end{cases}$
