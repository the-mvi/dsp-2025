## Основные теоретические положения

Под $y_n$ понимаем накопленную сумму интеграла до $n$-го шага интегрирования. \
Под $s_n$ понимаем значение подынтегральной функции в точке $x_n$.

**Формула прямоугольника** определяется как:
$$y_{n+1}=y_n + s_{n+\frac{1}{2}}, \: y_0=0$$

**Формула трапеций** определяется как:
$$y_{n+1}=y_n + \frac{1}{2}(s_n + s_{n+1}), \: y_0=0$$

**Формула Симпсона** определяется как:
$$y_{n+1}=y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}), \: y_0=0$$

**Формула для интегрирования по «правилу 3/8»** определяется как:
$$y_{n+2}=y_{n-1} + \frac{1}{8}(x_{n+2} + 3x_{n+1} + 3x_n + x_{n-1})$$

### Задание 1

Вывести формулы передаточных функций рекурсивных фильтров, соответствующих квадратурным формулам прямоугольников, трапеций и Симпсона. Построить графики передаточных функций и графики отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства полученных передаточных функций.

Рассмотрим **квадратурную формулу прямоугольников**:
$$y_{n+1} = y_n + s_{n+\frac{1}{2}}, y_0 = 0$$

Пусть $s_n = e^{i\omega n}$ при таком значении на выходе получится таже частота, что и на входе, но с другим масштабом/фазой: $y_n = H(\omega)e^{i\omega n}$.

Подставим следующее значение в рекурсию:
$$H(\omega)e^{i\omega (n+1)} = H(\omega)e^{i\omega n} + e^{i\omega (n+\frac{1}{2})}$$

$$H(\omega)e^{i\omega n}e^{i\omega} = H(\omega)e^{i\omega n} + e^{i\omega n}e^{i\omega \frac{1}{2}}$$

$$H(\omega)(e^{i\omega} - 1) = e^{i\omega \frac{1}{2}}$$

Отсюда получаем:

$$H(\omega) = \cfrac{e^{i\omega \frac{1}{2}}}{e^{i\omega} - 1}$$

Преобразуем:

$$H(\omega) = \cfrac{e^{i\omega \frac{1}{2}}}{e^{i\omega} - 1} \cdot \cfrac{e^{-i\omega \frac{1}{2}}}{e^{-i\omega \frac{1}{2}}} = \cfrac{1}{e^{i\omega \frac{1}{2}} - e^{-i\omega \frac{1}{2}}} = \cfrac{1}{2\cdot i \cdot sin(\frac{1}{2}\omega)}$$

Непрерывный интеграл, определяющий точное значение для $e^{i\omega t}$:
$$\int e^{i\omega t} dt = \cfrac{e^{i\omega t}}{i\omega}$$

Поймем насколько дискретная формула совпадает с идеалом при частоте $\omega$, то есть найдем отношение значений:
$$\gamma(\omega) = \cfrac{calculated}{exact} = \cfrac{i\omega \cdot e^{i\omega t}}{e^{i\omega t} \cdot 2 \cdot i \cdot sin(\frac{1}{2}\omega)} = \cfrac{i\omega}{2 \cdot i \cdot sin(\frac{1}{2}\omega)} = \cfrac{\omega}{2 \cdot sin(\frac{1}{2}\omega)}$$

Рассмотрим **квадратурную формулу трапеций**:
$$y_{n+1}=y_n + \frac{1}{2}(s_n + s_{n+1}), y_0 = 0$$

Возьмем аналогичные значения:
$$s_n = e^{i\omega n}, y_n = H(\omega)e^{i\omega n}$$

Подставим следующее значение в рекурсию:
$$H(\omega)e^{i\omega (n + 1)} = H(\omega)e^{i\omega n} + \frac{1}{2}(e^{i\omega n} + e^{i\omega (n+1)})$$
$$H(\omega)e^{i\omega n}e^{i\omega} = H(\omega)e^{i\omega n} + \frac{1}{2} e^{i\omega n} + \frac{1}{2} e^{i\omega n} e^{i\omega} $$
$$H(\omega)e^{i\omega} = H(\omega) + \frac{1}{2} e^{i\omega} + \frac{1}{2}$$
$$H(\omega)(e^{i\omega} - 1) = \frac{1}{2} ( e^{i\omega} +  1)$$

Получаем:
$$H(\omega) = \cfrac{e^{i\omega} +  1}{2(e^{i\omega} - 1)}$$

Преобразуем:
$$H(\omega) = H(\omega) \cdot \cfrac{e^{-i\omega \frac{1}{2}}}{e^{-i\omega \frac{1}{2}}} = \cfrac{cos(\frac{1}{2}\omega)}{2 \cdot i \cdot sin(\frac{1}{2}\omega)}$$

Непрерывный интеграл, определяющий точное значение для $e^{i\omega t}$:
$$\int e^{i\omega t} dt = \cfrac{e^{i\omega t}}{i\omega}$$

Поймем насколько дискретная формула совпадает с идеалом при частоте $\omega$, то есть найдем отношение значений:
$$\gamma(\omega) = \cfrac{calculated}{exact} = \cfrac{i\omega \cdot cos(\frac{1}{2}\omega) \cdot e^{i\omega t} }{2 \cdot e^{i\omega t} \cdot i \cdot sin(\frac{1}{2}\omega)} = \cfrac{\omega \cdot cos(\frac{1}{2}\omega)}{2 \cdot sin(\frac{1}{2}\omega)}$$

Рассмотрим **квадратурную формулу Симпсона**:
$$y_{n+1}=y_{n-1} + \frac{1}{3}(s_{n-1} + 4s_n + s_{n+1}), y_0=0$$

Возьмем аналогичные значения:
$$s_n = e^{i\omega n}, y_n = H(\omega)e^{i\omega n}$$

Подставим следующее значение в рекурсию:
$$H(\omega)e^{i\omega (n + 1)} = H(\omega)e^{i\omega (n-1)} + \frac{1}{3}(e^{i\omega (n-1)} + 4e^{i\omega n} + e^{i\omega (n+1)})$$
$$H(\omega)e^{i\omega n}e^{i\omega} = H(\omega)e^{i\omega n}e^{-i\omega} + \frac{1}{3}(e^{i\omega n}e^{-i\omega} + 4e^{i\omega n} + e^{i\omega n}e^{i\omega})$$
$$H(\omega)e^{i\omega} = H(\omega)e^{-i\omega} + \frac{1}{3}(e^{-i\omega} + e^{i\omega} + 4)$$
$$H(\omega)(e^{i\omega} - e^{-i\omega}) = \frac{1}{3}(e^{-i\omega} + e^{i\omega} + 4)$$

Получаем:
$$H(\omega) = \cfrac{e^{-i\omega} + e^{i\omega} + 4}{3(e^{i\omega} - e^{-i\omega})}$$

Преобразуем:
$$H(\omega) = \cfrac{2 + cos(\omega)}{3 \cdot i \cdot sin(\omega)} $$

Непрерывный интеграл, определяющий точное значение для $e^{i\omega t}$:
$$\int e^{i\omega t} dt = \cfrac{e^{i\omega t}}{i\omega}$$

Поймем насколько дискретная формула совпадает с идеалом при частоте $\omega$, то есть найдем отношение значений:
$$\gamma(\omega) = \cfrac{calculated}{exact} = \cfrac{i\omega \cdot (2 + cos(\omega)) \cdot e^{i\omega t} }{3 \cdot e^{i\omega t} \cdot i \cdot sin(\omega)}= \cfrac{\omega(2 + cos(\omega))}{3sin(\omega)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

eps = 1e-5
N = 5000
F = np.linspace(eps, 0.5 - eps, N, dtype=np.float64)
# W = 2 * pi * F
W = 2*np.pi*F
```

```python
H_rect = 1 / (2 * 1j * np.sin(0.5 * W))
H_trap = (np.cos(0.5 * W)) / (2 *1j * np.sin(0.5 * W))
H_simp = (2 + np.cos(W)) / (3 *1j * np.sin(W))
```

```python
plt.figure()
plt.plot(F, np.imag(H_rect), label="$H_{rect}$")
plt.plot(F, np.imag(H_trap), label="$H_{trap}$")
plt.plot(F, np.imag(H_simp), label="$H_{Simp}$")
plt.xlabel("f")
plt.ylabel("$\\tilde{H}(f)$")
plt.ylim((-5, 0.5))
plt.grid(True)
plt.legend()
plt.show()
```

По полученному графику можно увидеть, что все фильтры усиливают низкие частоты, а средние ослабевают в разной степени: значения передаточной функции по формуле трапеций в средней полосе лежат ближе к нулю — значит, они сильнее приглушают средние частоты; по формуле прямоугольника даёт умеренное ослабление; по формуле Симпсона чуть сильнее сохраняется средние частоты, чем у прямоугольника, высокие же частоты фильтры ослабляют, кроме Симпсона, который усиливает высокие частоты.

```python
G_rect = (W) / (2 * np.sin(0.5 * W))
G_trap = (W * np.cos(0.5 * W)) / (2 * np.sin(0.5 * W))
G_simp = (W * (2 + np.cos(W))) / (3 * np.sin(W))
```

```python
plt.figure()
plt.plot(F, G_rect, label="$\\gamma_{rect}$")
plt.plot(F, G_trap, label="$\\gamma_{trap}$")
plt.plot(F, G_simp, label="$\\gamma_{Simp}$")
plt.xlabel("f")
plt.grid(True)
plt.ylim((-0.5, 5))
plt.legend()
plt.show()
```

### Задание 2

Вывести формулу передаточной функции рекурсивного фильтра для интегрирования по «правилу 3/8». Построить график передаточной функции и график отношения вычисляемого в результате фильтрации значения к истинному. Проинтерпретировать частотные свойства передаточной функции.

Рассмотрим **формулу для интегрирования по «правилу 3/8»**:
$$y_{n+2}=y_{n-1} + \frac{1}{8}(x_{n+2} + 3x_{n+1} + 3x_n + x_{n-1})$$

Возьмем аналогичные значения:
$$x_n = e^{i\omega n}, y_n = H(\omega)e^{i\omega n}$$

Подставим следующее значение в рекурсию:
$$H(\omega)e^{i\omega (n + 2)} = H(\omega)e^{i\omega (n-1)} + \frac{1}{8}(e^{i\omega (n+2)} + 3e^{i\omega (n+1)} + 3e^{i\omega n} + e^{i\omega (n-1)})$$

$$H(\omega)e^{i\omega n}e^{2i\omega} = H(\omega)e^{i\omega n}e^{-i\omega} + \frac{1}{8}(e^{i\omega n}e^{2i\omega} + 3e^{i\omega n}e^{i\omega} + 3e^{i\omega n} + e^{i\omega n}e^{-i\omega})$$

$$H(\omega)e^{2i\omega} = H(\omega)e^{-i\omega} + \frac{1}{8}(e^{2i\omega} + 3e^{i\omega} + 3 + e^{-i\omega})$$

$$H(\omega)(e^{2i\omega} - e^{-i\omega})  = \frac{1}{8}(e^{2i\omega} + 3e^{i\omega} + 3 + e^{-i\omega})$$

Получаем:

$$H(\omega) = \cfrac{e^{2i\omega} + 3e^{i\omega} + 3 + e^{-i\omega}}{8(e^{2i\omega} - e^{-i\omega})}$$

Преобразуем:

$$H(\omega) = H(\omega) \cdot \cfrac{e^{-i\omega \frac{1}{2}}}{e^{-i\omega \frac{1}{2}}} = \cfrac{cos(\frac{3}{2}\omega) + 3cos(\frac{1}{2}\omega)}{8 \cdot i \cdot sin(\frac{3}{2}\omega)}$$

Непрерывный интеграл, определяющий точное значение для $e^{i\omega t}$:
$$\int e^{i\omega t} dt = \cfrac{e^{i\omega t}}{i\omega}$$

Поймем насколько дискретная формула совпадает с идеалом при частоте $\omega$, то есть найдем отношение значений:
$$\gamma(\omega) = \cfrac{calculated}{exact} = \cfrac{i\omega \cdot (cos(\frac{3}{2}\omega) + 3cos(\frac{1}{2}\omega)) \cdot e^{i\omega t} }{8 \cdot e^{i\omega t} \cdot i \cdot sin(\frac{3}{2}\omega)}= \cfrac{\omega \cdot (cos(\frac{3}{2}\omega) + 3cos(\frac{1}{2}\omega))}{8 \cdot sin(\frac{3}{2}\omega)} $$

```python
H_3_8 = (np.cos(3 * W / 2) + 3 * np.cos(W / 2)) / (8 * 1j * np.sin(3 * W / 2))
```

```python
plt.figure()
plt.plot(F, np.imag(H_3_8), label="$H_{3_8}$")
plt.xlabel("f")
plt.ylabel("$\\tilde{H}(f)$")
plt.ylim((-10, 10))
plt.grid(True)
plt.legend()
plt.show()
```

По данному графику видно, что в средней полосе, около 0.33, узкая «резонансная» зона, сигнал маскимально искажается. Также видно, низкие частоты проходят с большим усилением, средние подавляются сильнее, высокие — подавляются до нуля.

```python
G_3_8 = (W * (np.cos(3 * W / 2) + 3 * np.cos(W / 2))) / (8 * np.sin(3 * W / 2))
```

```python
plt.figure()
plt.plot(F, G_3_8, label="$\\gamma_{3_8}$")
plt.xlabel("f")
plt.grid(True)
plt.ylim((-10, 10))
plt.legend()
plt.show()
```
